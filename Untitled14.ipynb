```python
import os, numpy as np, pandas as pd
from pathlib import Path

# === パス設定（A: アップロード派）===
DATA_DIR = "."
FILES = [
    "living_kitchen0916_with_activity_full.csv",
    "sleepingroom0916_.csv",
    "washitsu0915_with_activity_full.csv",
]

# === （B: Drive派の場合の例）===
# DATA_DIR = "/content/drive/MyDrive/sensor_data"

def load_data(base_dir="."):
    dfs = []
    for name in FILES:
        fp = Path(base_dir) / name
        if not fp.exists():
            raise FileNotFoundError(f"Missing file: {fp}")
        df = pd.read_csv(fp, low_memory=False)
        if "timestamp" in df.columns:
            try:
                df["timestamp"] = pd.to_datetime(df["timestamp"])
            except Exception:
                pass
        dfs.append(df)
    data = pd.concat(dfs, axis=0, ignore_index=True, sort=False)
    for col in ["Occupied", "Activity", "Place"]:
        if col not in data.columns:
            data[col] = np.nan
    if "timestamp" in data.columns:
        data = data.sort_values("timestamp").reset_index(drop=True)
    return data

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
import numpy as np, pandas as pd

def build_preprocess(X):
    cat_cols = [c for c in X.columns if X[c].dtype == "object"]
    num_cols = [c for c in X.columns if c not in cat_cols]
    num_pipe = Pipeline([("imputer", SimpleImputer(strategy="median")),
                         ("scaler", StandardScaler())])
    cat_pipe = Pipeline([("imputer", SimpleImputer(strategy="most_frequent")),
                         ("onehot", OneHotEncoder(handle_unknown="ignore"))])
    return ColumnTransformer([("num", num_pipe, num_cols),
                              ("cat", cat_pipe, cat_cols)])

from sklearn.model_selection import train_test_split

def time_split(X, y, test_size=0.2):
    # まず時系列分割
    Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=test_size, shuffle=False)
    if len(pd.Series(ytr).dropna().unique()) < 2 or len(pd.Series(yte).dropna().unique()) < 2:
        # 層別シャッフルで再挑戦
        mask = y.notna()
        Xv, yv = X[mask], y[mask]
        try:
            Xtr, Xte, ytr, yte = train_test_split(Xv, yv, test_size=test_size, shuffle=True, stratify=yv)
        except Exception:
            # それでもダメなら、そのまま返す（呼び出し側でスキップ判断）
            pass
    return Xtr, Xte, ytr, yte


def eval_bin(y_true, y_pred, y_prob=None):
    y_true = pd.Series(y_true).astype(int)
    y_pred = pd.Series(y_pred).astype(int)
    acc = accuracy_score(y_true, y_pred)
    pre, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    out = {"accuracy": float(acc), "precision": float(pre), "recall": float(rec), "f1": float(f1)}
    if y_prob is not None and len(np.unique(y_true)) == 2:
        try:
            out["auc"] = float(roc_auc_score(y_true, y_prob))
        except Exception:
            pass
    return out

def build_sequences(X_arr, y_ser, window=30, step=1):
    Xs, ys = [], []
    for i in range(window, len(X_arr), step):
        Xs.append(X_arr[i-window:i, :])
        ys.append(int(y_ser.iloc[i]))
    return np.array(Xs), np.array(ys)

# ==== 実行 ====
out_dir = Path("model_outputs"); out_dir.mkdir(exist_ok=True, parents=True)
data = load_data(DATA_DIR)
data = data[data["Occupied"].notna()].copy()
data["Occupied"] = data["Occupied"].astype(int)

exclude = ["timestamp", "Place", "Activity", "Occupied", "Number of People"]
feat_cols = [c for c in data.columns if c not in exclude]
X = data[feat_cols].copy()
y = data["Occupied"].astype(int)

# 前処理 & 分割
preprocess = build_preprocess(X)
Xtr, Xte, ytr, yte = time_split(X, y, 0.2)

# --- RF ---
rf = Pipeline([("prep", preprocess),
               ("clf", RandomForestClassifier(n_estimators=160, random_state=42, n_jobs=-1))])
rf.fit(Xtr, ytr)
rf_pred = rf.predict(Xte)
try:
    rf_prob = rf.predict_proba(Xte)[:,1]
except Exception:
    rf_prob = None
rf_m = {"model":"RandomForest", **eval_bin(yte, rf_pred, rf_prob)}

# --- HGB (XGBoost相当) ---
hgb = Pipeline([("prep", preprocess),
                ("clf", HistGradientBoostingClassifier(learning_rate=0.1, random_state=42))])
hgb.fit(Xtr, ytr)
hgb_pred = hgb.predict(Xte)
try:
    from scipy.special import expit
    hgb_prob = expit(hgb.decision_function(Xte))
except Exception:
    hgb_prob = None
hgb_m = {"model":"SklearnHGB(XGB-like)", **eval_bin(yte, hgb_pred, hgb_prob)}

# --- LSTM / TCN 用の系列データ ---
Xtr_mat = preprocess.fit_transform(Xtr)
X_all_mat = preprocess.transform(X)
if hasattr(X_all_mat, "toarray"):
    X_all_mat = X_all_mat.toarray()

train_end = len(Xtr)
WINDOW = 30  # 30ステップ窓（必要に応じて調整）
X_tr_seq, y_tr_seq = build_sequences(X_all_mat[:train_end], y.iloc[:train_end], window=WINDOW)
X_te_seq, y_te_seq = build_sequences(X_all_mat[train_end:], y.iloc[train_end:], window=WINDOW)

# データが多ければ上限キャップ（学習時間短縮）
def cap(Xs, ys, mx=8000):
    if len(Xs) > mx:
        idx = np.linspace(0, len(Xs)-1, mx).astype(int)
        return Xs[idx], ys[idx]
    return Xs, ys
X_tr_seq, y_tr_seq = cap(X_tr_seq, y_tr_seq)
X_te_seq, y_te_seq = cap(X_te_seq, y_te_seq)

# --- LSTM ---
lstm_m = {"model": "LSTM", "note": "skipped"}
try:
    import tensorflow as tf
    from tensorflow import keras
    from tensorflow.keras import layers
    if len(X_tr_seq)>100 and len(np.unique(y_tr_seq))==2 and len(np.unique(y_te_seq))==2:
        input_shape = (X_tr_seq.shape[1], X_tr_seq.shape[2])
        lstm = keras.Sequential([
            layers.Input(shape=input_shape),
            layers.Masking(mask_value=0.0),
            layers.LSTM(64),
            layers.Dense(32, activation="relu"),
            layers.Dense(1, activation="sigmoid")
        ])
        lstm.compile(optimizer=keras.optimizers.Adam(1e-3), loss="binary_crossentropy")
        lstm.fit(X_tr_seq, y_tr_seq, epochs=6, batch_size=128, verbose=0, validation_split=0.1)
        y_prob_lstm = lstm.predict(X_te_seq, verbose=0).ravel()
        y_pred_lstm = (y_prob_lstm >= 0.5).astype(int)
        lstm_m = {"model":"LSTM", **eval_bin(y_te_seq, y_pred_lstm, y_prob_lstm)}
except Exception as e:
    print("LSTM skipped:", e)

# --- TCN（因果Dilated Conv） ---
tcn_m = {"model": "TCN", "note": "skipped"}
try:
    import tensorflow as tf
    from tensorflow import keras
    from tensorflow.keras import layers
    if len(X_tr_seq)>100 and len(np.unique(y_tr_seq))==2 and len(np.unique(y_te_seq))==2:
        inp = layers.Input(shape=(X_tr_seq.shape[1], X_tr_seq.shape[2]))
        x = inp
        for d in [1, 2, 4, 8]:
            x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
            x = layers.ReLU()(x)
            x = layers.Dropout(0.1)(x)
        x = layers.GlobalAveragePooling1D()(x)
        x = layers.Dense(32, activation="relu")(x)
        out = layers.Dense(1, activation="sigmoid")(x)
        tcn = keras.Model(inp, out)
        tcn.compile(optimizer=keras.optimizers.Adam(1e-3), loss="binary_crossentropy")
        tcn.fit(X_tr_seq, y_tr_seq, epochs=6, batch_size=128, verbose=0, validation_split=0.1)
        y_prob_tcn = tcn.predict(X_te_seq, verbose=0).ravel()
        y_pred_tcn = (y_prob_tcn >= 0.5).astype(int)
        tcn_m = {"model":"TCN", **eval_bin(y_te_seq, y_pred_tcn, y_prob_tcn)}
except Exception as e:
    print("TCN skipped:", e)

# まとめ & 保存
res = pd.DataFrame([rf_m, hgb_m, lstm_m, tcn_m])
res.to_csv(out_dir / "occupied_four_models_comparison.csv", index=False)
res
```

```python
# === 設定 ===
DATA_DIR = "."  # 3つのCSVを置いたパス（Driveなら "/content/drive/MyDrive/sensor_data" 等）
FILES = [
    "living_kitchen0916_with_activity_full.csv",
    "sleepingroom0916_.csv",
    "washitsu0915_with_activity_full.csv",
]
GROUP_BY = "source_file"   # "source_file" or "Place"
COVERAGE_MIN = 0.30        # 学習用trainにおける非欠損率の下限（列除去の閾値）
WINDOW = 60                # LSTM/TCNの窓長（ステップ数）
MAX_ROWS_PER_GROUP = 40000 # 各グループでの最大行数（速度とメモリ対策）
NN_EPOCHS = 8
NN_BATCH = 128

# === ここから実装 ===
import os, numpy as np, pandas as pd
from pathlib import Path

def load_data(base_dir, files):
    dfs = []
    for name in files:
        fp = Path(base_dir) / name
        if not fp.exists():
            raise FileNotFoundError(f"Missing file: {fp}")
        df = pd.read_csv(fp, low_memory=False)
        if "timestamp" in df.columns:
            try:
                df["timestamp"] = pd.to_datetime(df["timestamp"])
            except Exception:
                pass
        df["source_file"] = name
        dfs.append(df)
    data = pd.concat(dfs, axis=0, ignore_index=True, sort=False)
    for col in ["Occupied", "Activity", "Place"]:
        if col not in data.columns:
            data[col] = np.nan
    if "timestamp" in data.columns:
        data = data.sort_values("timestamp").reset_index(drop=True)
    return data

from sklearn.model_selection import train_test_split
def time_split(X, y, test_size=0.2):
    Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=test_size, shuffle=False)
    if len(pd.Series(yte).dropna().unique()) < 2:
        mask = y.notna()
        Xv, yv = X[mask], y[mask]
        Xtr, Xte, ytr, yte = train_test_split(Xv, yv, test_size=test_size, shuffle=True, stratify=yv)
    return Xtr, Xte, ytr, yte

from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score, f1_score
def eval_bin(y_true, y_pred, y_prob=None):
    y_true = pd.Series(y_true).astype(int)
    y_pred = pd.Series(y_pred).astype(int)
    acc = accuracy_score(y_true, y_pred)
    pre, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    out = {"accuracy": float(acc), "precision": float(pre), "recall": float(rec), "f1": float(f1)}
    if y_prob is not None and len(np.unique(y_true)) == 2:
        try:
            out["auc"] = float(roc_auc_score(y_true, y_prob))
        except Exception:
            pass
    return out

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

def build_preprocess(X):
    cat_cols = [c for c in X.columns if X[c].dtype == "object"]
    num_cols = [c for c in X.columns if c not in cat_cols]
    num_pipe = Pipeline([("imputer", SimpleImputer(strategy="median")),
                         ("scaler", StandardScaler())])
    cat_pipe = Pipeline([("imputer", SimpleImputer(strategy="most_frequent")),
                         ("onehot", OneHotEncoder(handle_unknown="ignore"))])
    return ColumnTransformer([("num", num_pipe, num_cols),
                              ("cat", cat_pipe, cat_cols)])

def select_by_coverage(Xtr, min_ratio=COVERAGE_MIN):
    cov = Xtr.notna().mean()
    keep = cov[cov >= min_ratio].index.tolist()
    return keep

def build_sequences(X_arr, y_ser, window=WINDOW, step=1):
    Xs, ys = [], []
    for i in range(window, len(X_arr), step):
        Xs.append(X_arr[i-window:i, :])
        ys.append(int(y_ser.iloc[i]))
    return np.array(Xs), np.array(ys)

def cap_samples(Xs, ys, mx=8000):
    if len(Xs) > mx:
        idx = np.linspace(0, len(Xs)-1, mx).astype(int)
        return Xs[idx], ys[idx]
    return Xs, ys

from sklearn.utils.class_weight import compute_class_weight

def compute_cw(y):
    y = pd.Series(y).astype(int)
    classes = np.sort(y.unique())
    if len(classes) < 2:
        # 単一クラス → class_weightは使わない
        return None
    cw = compute_class_weight(class_weight="balanced", classes=classes, y=y)
    return {int(c): float(w) for c, w in zip(classes, cw)}


from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from scipy.special import expit

def run_group(df_group, group_name, out_dir):
    # 1) ラベル整備
    df_group = df_group[df_group["Occupied"].notna()].copy()
    if len(df_group) < 200:
        return pd.DataFrame([{"group": group_name, "model":"N/A", "note":"too few rows"}])

    df_group["Occupied"] = df_group["Occupied"].astype(int)

    # 2) 特徴量
    exclude = ["timestamp", "Place", "Activity", "Occupied", "Number of People", "source_file"]
    feat_cols = [c for c in df_group.columns if c not in exclude]
    X = df_group[feat_cols].copy()
    y = df_group["Occupied"].astype(int)

    # 全体が単一クラスなら、このグループは評価不能
    if len(y.unique()) < 2:
        return pd.DataFrame([{"group": group_name, "model":"N/A", "note":"only one class in this group"}])

    # 3) 分割
    Xtr, Xte, ytr, yte = time_split(X, y, 0.2)

    # 分割後に依然として片側が単一クラスの場合はスキップ
    if len(pd.Series(ytr).unique()) < 2 or len(pd.Series(yte).unique()) < 2:
        return pd.DataFrame([{"group": group_name, "model":"N/A", "note":"train/test still single-class after split"}])

    # 4) 低カバレッジ列選別（trainベース）
    keep_cols = select_by_coverage(Xtr, COVERAGE_MIN)
    if len(keep_cols) == 0:
        return pd.DataFrame([{"group": group_name, "model":"N/A", "note":"no columns after coverage filter"}])
    Xtr, Xte = Xtr[keep_cols], Xte[keep_cols]

    # 5) 前処理器
    preprocess = build_preprocess(Xtr)

    # 6) クラス重み（RF/HGB用）
    cw = compute_cw(ytr)  # None の可能性あり

    results = []

    # === RandomForest ===
    rf = Pipeline([
        ("prep", preprocess),
        ("clf", RandomForestClassifier(
            n_estimators=220,
            class_weight=("balanced" if cw is not None else None),
            random_state=42, n_jobs=-1
        ))
    ])
    rf.fit(Xtr, ytr)
    rf_pred = rf.predict(Xte)
    try:
        rf_prob = rf.predict_proba(Xte)[:, 1]
    except Exception:
        rf_prob = None
    results.append({"group": group_name, "model":"RandomForest", **eval_bin(yte, rf_pred, rf_prob)})

    # === HGB (XGB相当) ===
    # preprocessを train に fit → train/test transform（リーク防止）
    preprocess.fit(Xtr)
    Xtr_mat = preprocess.transform(Xtr); Xte_mat = preprocess.transform(Xte)
    if hasattr(Xtr_mat, "toarray"):
        Xtr_mat = Xtr_mat.toarray(); Xte_mat = Xte_mat.toarray()

    hgb = HistGradientBoostingClassifier(learning_rate=0.1, random_state=42)
    # sample_weight を安全に生成（cwがNoneなら等重み）
    if cw is not None:
        w = pd.Series(ytr).map(lambda v: cw.get(int(v), 1.0)).values
        hgb.fit(Xtr_mat, ytr, sample_weight=w)
    else:
        hgb.fit(Xtr_mat, ytr)
    hgb_pred = hgb.predict(Xte_mat)
    try:
        hgb_prob = expit(hgb.decision_function(Xte_mat))
    except Exception:
        hgb_prob = None
    results.append({"group": group_name, "model":"SklearnHGB(XGB-like)", **eval_bin(yte, hgb_pred, hgb_prob)})

    # === NN (LSTM/TCN) ===
    lstm_m = {"group": group_name, "model":"LSTM", "note":"skipped"}
    tcn_m  = {"group": group_name, "model":"TCN",  "note":"skipped"}
    try:
        import tensorflow as tf
        from tensorflow import keras
        from tensorflow.keras import layers
        from tensorflow.keras.callbacks import EarlyStopping

        # グループ全体を同じ前処理で変換（trainでfit済みのpreprocessを使用）
        X_all_mat = preprocess.transform(df_group[keep_cols])
        if hasattr(X_all_mat, "toarray"):
            X_all_mat = X_all_mat.toarray()

        train_end = len(Xtr)
        X_tr_seq, y_tr_seq = build_sequences(X_all_mat[:train_end], y.iloc[:train_end], window=WINDOW, step=1)
        X_te_seq, y_te_seq = build_sequences(X_all_mat[train_end:], y.iloc[train_end:], window=WINDOW, step=1)
        X_tr_seq, y_tr_seq = cap_samples(X_tr_seq, y_tr_seq, mx=8000)
        X_te_seq, y_te_seq = cap_samples(X_te_seq, y_te_seq, mx=8000)

        # NN用class_weight（系列yから計算）
        cw_nn = compute_cw(pd.Series(y_tr_seq))  # None 可
        callbacks = [EarlyStopping(monitor="val_loss", patience=2, restore_best_weights=True)]

        # 学習に十分な条件を満たすときのみ実行
        if len(X_tr_seq) > 100 and len(np.unique(y_tr_seq)) == 2 and len(np.unique(y_te_seq)) == 2:
            input_shape = (X_tr_seq.shape[1], X_tr_seq.shape[2])

            # LSTM
            lstm = keras.Sequential([
                layers.Input(shape=input_shape),
                layers.Masking(mask_value=0.0),
                layers.LSTM(64),
                layers.Dropout(0.2),
                layers.Dense(32, activation="relu"),
                layers.Dense(1, activation="sigmoid")
            ])
            lstm.compile(optimizer=keras.optimizers.Adam(1e-3), loss="binary_crossentropy")
            lstm.fit(
                X_tr_seq, y_tr_seq,
                epochs=NN_EPOCHS, batch_size=NN_BATCH, validation_split=0.1,
                class_weight=cw_nn, callbacks=callbacks, verbose=0
            )
            y_prob_lstm = lstm.predict(X_te_seq, verbose=0).ravel()
            y_pred_lstm = (y_prob_lstm >= 0.5).astype(int)
            lstm_m = {"group": group_name, "model":"LSTM", **eval_bin(y_te_seq, y_pred_lstm, y_prob_lstm)}

            # TCN（因果ダイレート）
            inp = layers.Input(shape=input_shape)
            x = inp
            for d in [1, 2, 4, 8]:
                x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
                x = layers.ReLU()(x)
                x = layers.Dropout(0.2)(x)
            x = layers.GlobalAveragePooling1D()(x)
            x = layers.Dense(32, activation="relu")(x)
            out = layers.Dense(1, activation="sigmoid")(x)
            tcn = keras.Model(inp, out)
            tcn.compile(optimizer=keras.optimizers.Adam(1e-3), loss="binary_crossentropy")
            tcn.fit(
                X_tr_seq, y_tr_seq,
                epochs=NN_EPOCHS, batch_size=NN_BATCH, validation_split=0.1,
                class_weight=cw_nn, callbacks=callbacks, verbose=0
            )
            y_prob_tcn = tcn.predict(X_te_seq, verbose=0).ravel()
            y_pred_tcn = (y_prob_tcn >= 0.5).astype(int)
            tcn_m = {"group": group_name, "model":"TCN", **eval_bin(y_te_seq, y_pred_tcn, y_prob_tcn)}
    except Exception as e:
        print(f"[{group_name}] NN skipped:", e)

    results.extend([lstm_m, tcn_m])

    df_res = pd.DataFrame(results)
    out_dir.mkdir(parents=True, exist_ok=True)
    df_res.to_csv(out_dir / f"occupied_models_{group_name}.csv", index=False)
    return df_res


# === 実行 ===
data = load_data(DATA_DIR, FILES)
if GROUP_BY not in data.columns:
    raise ValueError(f"GROUP_BY='{GROUP_BY}' がデータに見つかりません。columns: {list(data.columns)[:10]} ...")

groups = data[GROUP_BY].fillna("unknown").unique().tolist()
print("Groups:", groups)

all_res = []
out_dir = Path("model_outputs")
for g in groups:
    sub = data[data[GROUP_BY]==g].copy()
    res_g = run_group(sub, str(g), out_dir)
    all_res.append(res_g)

all_res = pd.concat(all_res, axis=0, ignore_index=True)
all_res.to_csv(out_dir / "occupied_models_all_groups.csv", index=False)

# 集計（macro平均）
agg = (all_res.groupby("model")[["accuracy","precision","recall","f1","auc"]]
       .mean(numeric_only=True)
       .reset_index())
agg.to_csv(out_dir / "occupied_models_group_macro_avg.csv", index=False)
display(agg)
print("Saved per-group CSVs to:", out_dir)
```

```python
# ==================== 設定 ====================
DATA_DIR = "."  # 3つのCSVを置いたパス（例: "/content/drive/MyDrive/sensor_data"）
FILES = [
    "living_kitchen0916_with_activity_full.csv",
    "sleepingroom0916_.csv",
    "washitsu0915_with_activity_full.csv",
]
GROUP_BY = "source_file"   # "source_file" か "Place" を選択
COVERAGE_MIN = 0.30        # 学習(train)内の非欠損率がこれ未満の列は除外
WINDOW = 60                # LSTM/TCN の系列窓長（ステップ数）
MAX_ROWS_PER_GROUP = 40000 # 各グループの最大行数（速度・メモリ対策）
NN_EPOCHS = 8
NN_BATCH = 128
DEV_RATIO = 0.10           # 学習データ末尾の一定割合を dev として閾値最適化に使用

LEAKY_NAME_PATTERNS = ["occupied", "presence", "activity"]  # 列名に含まれたら除外（小文字比較）
CORR_DROP_TH = 0.98        # 学習データで |corr(feature, Occupied)| > しきい値 の列を除外

# ==================== 実装 ====================
import os, numpy as np, pandas as pd
from pathlib import Path

def load_data(base_dir, files):
    dfs = []
    for name in files:
        fp = Path(base_dir) / name
        if not fp.exists():
            raise FileNotFoundError(f"Missing file: {fp}")
        df = pd.read_csv(fp, low_memory=False)
        if "timestamp" in df.columns:
            try:
                df["timestamp"] = pd.to_datetime(df["timestamp"])
            except Exception:
                pass
        df["source_file"] = name
        dfs.append(df)
    data = pd.concat(dfs, axis=0, ignore_index=True, sort=False)
    for col in ["Occupied", "Activity", "Place"]:
        if col not in data.columns:
            data[col] = np.nan
    if "timestamp" in data.columns:
        data = data.sort_values("timestamp").reset_index(drop=True)
    return data

from sklearn.model_selection import train_test_split
def time_split(X, y, test_size=0.2):
    # 時系列でまず分割
    Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=test_size, shuffle=False)
    # どちらかが単一クラスなら層別シャッフルにフォールバック
    if len(pd.Series(ytr).dropna().unique()) < 2 or len(pd.Series(yte).dropna().unique()) < 2:
        mask = y.notna()
        Xv, yv = X[mask], y[mask]
        try:
            Xtr, Xte, ytr, yte = train_test_split(Xv, yv, test_size=test_size, shuffle=True, stratify=yv)
        except Exception:
            pass
    return Xtr, Xte, ytr, yte

from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score, f1_score
def eval_bin(y_true, y_pred, y_prob=None):
    y_true = pd.Series(y_true).astype(int)
    y_pred = pd.Series(y_pred).astype(int)
    acc = accuracy_score(y_true, y_pred)
    pre, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    out = {"accuracy": float(acc), "precision": float(pre), "recall": float(rec), "f1": float(f1)}
    if y_prob is not None and len(np.unique(y_true)) == 2:
        try:
            out["auc"] = float(roc_auc_score(y_true, y_prob))
        except Exception:
            pass
    return out

def find_best_threshold(y_true, y_prob, grid=np.linspace(0.1, 0.9, 17)):
    y_true = pd.Series(y_true).astype(int)
    best_thr, best_f1 = 0.5, -1.0
    for thr in grid:
        f1 = f1_score(y_true, (y_prob >= thr).astype(int), zero_division=0)
        if f1 > best_f1:
            best_f1, best_thr = f1, thr
    return float(best_thr), float(best_f1)

# 前処理（OneHot + 標準化）
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
def build_preprocess(X):
    cat_cols = [c for c in X.columns if X[c].dtype == "object"]
    num_cols = [c for c in X.columns if c not in cat_cols]
    num_pipe = Pipeline([("imputer", SimpleImputer(strategy="median")),
                         ("scaler", StandardScaler())])
    cat_pipe = Pipeline([("imputer", SimpleImputer(strategy="most_frequent")),
                         ("onehot", OneHotEncoder(handle_unknown="ignore"))])
    return ColumnTransformer([("num", num_pipe, num_cols),
                              ("cat", cat_pipe, cat_cols)])

# 低カバレッジ列
def select_by_coverage(Xtr, min_ratio=COVERAGE_MIN):
    cov = Xtr.notna().mean()
    keep = cov[cov >= min_ratio].index.tolist()
    return keep

# リーク防止（名前）
def drop_leaky_by_name(cols):
    keep = []
    for c in cols:
        lc = c.lower()
        if any(p in lc for p in LEAKY_NAME_PATTERNS):
            continue
        keep.append(c)
    return keep

# リーク防止（相関）
def drop_leaky_by_corr(Xtr_df, ytr_ser, cols):
    num_cols = [c for c in cols if pd.api.types.is_numeric_dtype(Xtr_df[c])]
    if len(num_cols) == 0:
        return cols
    corr = pd.concat([Xtr_df[num_cols], ytr_ser.reset_index(drop=True)], axis=1).corr(numeric_only=True)
    ycorr = corr[ytr_ser.name].drop(labels=[ytr_ser.name], errors="ignore")
    to_drop = set(ycorr[ycorr.abs() > CORR_DROP_TH].index.tolist())
    return [c for c in cols if c not in to_drop]

# シーケンス化
def build_sequences(X_arr, y_ser, window=WINDOW, step=1):
    Xs, ys = [], []
    for i in range(window, len(X_arr), step):
        Xs.append(X_arr[i-window:i, :])
        ys.append(int(y_ser.iloc[i]))
    return np.array(Xs), np.array(ys)

def cap_samples(Xs, ys, mx=8000):
    if len(Xs) > mx:
        idx = np.linspace(0, len(Xs)-1, mx).astype(int)
        return Xs[idx], ys[idx]
    return Xs, ys

# クラス重み
from sklearn.utils.class_weight import compute_class_weight
def compute_cw(y):
    y = pd.Series(y).astype(int)
    classes = np.sort(y.unique())
    if len(classes) < 2:
        return None
    cw = compute_class_weight(class_weight="balanced", classes=classes, y=y)
    return {int(c): float(w) for c, w in zip(classes, cw)}

# モデル
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from scipy.special import expit

def run_group(df_group, group_name, out_dir):
    # 1) ラベル整備
    df_group = df_group[df_group["Occupied"].notna()].copy()
    if len(df_group) < 200:
        return pd.DataFrame([{"group": group_name, "model":"N/A", "note":"too few rows"}])
    df_group["Occupied"] = df_group["Occupied"].astype(int)

    # 全体が単一クラスなら終了
    if len(df_group["Occupied"].unique()) < 2:
        return pd.DataFrame([{"group": group_name, "model":"N/A", "note":"only one class in this group"}])

    # 速度対策
    if len(df_group) > MAX_ROWS_PER_GROUP:
        idx = np.linspace(0, len(df_group)-1, MAX_ROWS_PER_GROUP).astype(int)
        df_group = df_group.iloc[idx].reset_index(drop=True)

    # 2) 特徴量
    exclude = ["timestamp", "Place", "Activity", "Occupied", "Number of People", "source_file"]
    feat_cols = [c for c in df_group.columns if c not in exclude]
    X = df_group[feat_cols].copy()
    y = df_group["Occupied"].astype(int)

    # 3) 分割
    Xtr, Xte, ytr, yte = time_split(X, y, 0.2)
    if len(pd.Series(ytr).unique()) < 2 or len(pd.Series(yte).unique()) < 2:
        return pd.DataFrame([{"group": group_name, "model":"N/A", "note":"train/test single-class after split"}])

    # 4) 低カバレッジ → リーク除去（名前 & 相関）
    keep_cols = select_by_coverage(Xtr, COVERAGE_MIN)
    keep_cols = drop_leaky_by_name(keep_cols)
    keep_cols = drop_leaky_by_corr(Xtr[keep_cols], ytr, keep_cols)
    if len(keep_cols) == 0:
        return pd.DataFrame([{"group": group_name, "model":"N/A", "note":"no columns after filters"}])

    Xtr, Xte = Xtr[keep_cols], Xte[keep_cols]

    # 5) 前処理
    preprocess = build_preprocess(Xtr)

    # 6) RF/HGB 用の dev 分割（学習末尾の一部で閾値最適化）
    dev_n = max(int(len(Xtr) * DEV_RATIO), 100)
    if len(Xtr) <= dev_n + 100:
        dev_n = max(50, int(len(Xtr) * 0.1))  # 最低限の調整
    Xtr_tr, Xtr_dev = Xtr.iloc[:-dev_n], Xtr.iloc[-dev_n:]
    ytr_tr, ytr_dev = ytr.iloc[:-dev_n], ytr.iloc[-dev_n:]

    cw = compute_cw(ytr_tr)  # None 可

    results = []

    # === RandomForest ===
    rf = Pipeline([
        ("prep", preprocess),
        ("clf", RandomForestClassifier(
            n_estimators=220,
            class_weight=("balanced" if cw is not None else None),
            random_state=42, n_jobs=-1
        ))
    ])
    rf.fit(Xtr_tr, ytr_tr)
    rf_prob_dev = rf.predict_proba(Xtr_dev)[:, 1]
    thr_rf, _ = find_best_threshold(ytr_dev, rf_prob_dev)

    rf_prob = rf.predict_proba(Xte)[:, 1]
    rf_pred = (rf_prob >= thr_rf).astype(int)
    results.append({"group": group_name, "model":"RandomForest", **eval_bin(yte, rf_pred, rf_prob)})

    # === HGB (XGB相当) ===
    preprocess.fit(Xtr_tr)
    Xtr_tr_mat = preprocess.transform(Xtr_tr); Xtr_dev_mat = preprocess.transform(Xtr_dev); Xte_mat = preprocess.transform(Xte)
    if hasattr(Xtr_tr_mat, "toarray"):
        Xtr_tr_mat = Xtr_tr_mat.toarray()
        Xtr_dev_mat = Xtr_dev_mat.toarray()
        Xte_mat = Xte_mat.toarray()

    hgb = HistGradientBoostingClassifier(learning_rate=0.1, random_state=42)
    if cw is not None:
        w_tr = pd.Series(ytr_tr).map(lambda v: cw.get(int(v), 1.0)).values
        hgb.fit(Xtr_tr_mat, ytr_tr, sample_weight=w_tr)
    else:
        hgb.fit(Xtr_tr_mat, ytr_tr)

    hgb_prob_dev = expit(hgb.decision_function(Xtr_dev_mat))
    thr_hgb, _ = find_best_threshold(ytr_dev, hgb_prob_dev)

    hgb_prob = expit(hgb.decision_function(Xte_mat))
    hgb_pred = (hgb_prob >= thr_hgb).astype(int)
    results.append({"group": group_name, "model":"SklearnHGB(XGB-like)", **eval_bin(yte, hgb_pred, hgb_prob)})

    # === NN (LSTM/TCN) ===
    lstm_m = {"group": group_name, "model":"LSTM", "note":"skipped"}
    tcn_m  = {"group": group_name, "model":"TCN",  "note":"skipped"}
    try:
        import tensorflow as tf
        from tensorflow import keras
        from tensorflow.keras import layers
        from tensorflow.keras.callbacks import EarlyStopping

        # グループ全体を同じ前処理で変換（Xtr_trにfit済み）
        X_all_mat = preprocess.transform(df_group[keep_cols])
        if hasattr(X_all_mat, "toarray"):
            X_all_mat = X_all_mat.toarray()

        # 学習区間＝Xtrの長さ（dev 部分は後ろの dev_n シーケンスで）
        train_end = len(Xtr)
        X_tr_seq, y_tr_seq = build_sequences(X_all_mat[:train_end], y.iloc[:train_end], window=WINDOW, step=1)
        X_te_seq, y_te_seq = build_sequences(X_all_mat[train_end:], y.iloc[train_end:], window=WINDOW, step=1)

        # dev分割（系列にも対応）
        dev_n_seq = max(int(len(X_tr_seq) * DEV_RATIO), 100)
        if len(X_tr_seq) <= dev_n_seq + 100:
            dev_n_seq = max(50, int(len(X_tr_seq) * 0.1))
        Xtr_seq_tr, Xtr_seq_dev = X_tr_seq[:-dev_n_seq], X_tr_seq[-dev_n_seq:]
        ytr_seq_tr, ytr_seq_dev = y_tr_seq[:-dev_n_seq], y_tr_seq[-dev_n_seq:]

        # サンプル数制限
        Xtr_seq_tr, ytr_seq_tr = cap_samples(Xtr_seq_tr, ytr_seq_tr, mx=8000)
        Xtr_seq_dev, ytr_seq_dev = cap_samples(Xtr_seq_dev, ytr_seq_dev, mx=2000)
        X_te_seq, y_te_seq = cap_samples(X_te_seq, y_te_seq, mx=8000)

        # 実行条件を満たすか確認
        if len(Xtr_seq_tr) > 100 and len(np.unique(ytr_seq_tr)) == 2 and len(np.unique(y_te_seq)) == 2:
            input_shape = (Xtr_seq_tr.shape[1], Xtr_seq_tr.shape[2])
            cw_nn = compute_cw(pd.Series(ytr_seq_tr))  # None 可
            callbacks = [EarlyStopping(monitor="val_loss", patience=2, restore_best_weights=True)]

            # --- LSTM ---
            lstm = keras.Sequential([
                layers.Input(shape=input_shape),
                layers.Masking(mask_value=0.0),
                layers.LSTM(64),
                layers.Dropout(0.2),
                layers.Dense(32, activation="relu"),
                layers.Dense(1, activation="sigmoid")
            ])
            lstm.compile(optimizer=keras.optimizers.Adam(1e-3), loss="binary_crossentropy")
            lstm.fit(
                Xtr_seq_tr, ytr_seq_tr,
                epochs=NN_EPOCHS, batch_size=NN_BATCH, validation_data=(Xtr_seq_dev, ytr_seq_dev),
                class_weight=cw_nn, callbacks=callbacks, verbose=0
            )
            y_prob_lstm_dev = lstm.predict(Xtr_seq_dev, verbose=0).ravel()
            thr_lstm, _ = find_best_threshold(ytr_seq_dev, y_prob_lstm_dev)
            y_prob_lstm = lstm.predict(X_te_seq, verbose=0).ravel()
            y_pred_lstm = (y_prob_lstm >= thr_lstm).astype(int)
            lstm_m = {"group": group_name, "model":"LSTM", **eval_bin(y_te_seq, y_pred_lstm, y_prob_lstm)}

            # --- TCN（因果 Dilated Conv）---
            inp = layers.Input(shape=input_shape)
            x = inp
            for d in [1, 2, 4, 8]:
                x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
                x = layers.ReLU()(x)
                x = layers.Dropout(0.2)(x)
            x = layers.GlobalAveragePooling1D()(x)
            x = layers.Dense(32, activation="relu")(x)
            out = layers.Dense(1, activation="sigmoid")(x)
            tcn = keras.Model(inp, out)
            tcn.compile(optimizer=keras.optimizers.Adam(1e-3), loss="binary_crossentropy")
            tcn.fit(
                Xtr_seq_tr, ytr_seq_tr,
                epochs=NN_EPOCHS, batch_size=NN_BATCH, validation_data=(Xtr_seq_dev, ytr_seq_dev),
                class_weight=cw_nn, callbacks=callbacks, verbose=0
            )
            y_prob_tcn_dev = tcn.predict(Xtr_seq_dev, verbose=0).ravel()
            thr_tcn, _ = find_best_threshold(ytr_seq_dev, y_prob_tcn_dev)
            y_prob_tcn = tcn.predict(X_te_seq, verbose=0).ravel()
            y_pred_tcn = (y_prob_tcn >= thr_tcn).astype(int)
            tcn_m = {"group": group_name, "model":"TCN", **eval_bin(y_te_seq, y_pred_tcn, y_prob_tcn)}
    except Exception as e:
        print(f"[{group_name}] NN skipped:", e)

    results.extend([lstm_m, tcn_m])

    df_res = pd.DataFrame(results)
    out_dir.mkdir(parents=True, exist_ok=True)
    df_res.to_csv(out_dir / f"occupied_models_{group_name}.csv", index=False)
    return df_res

# ==================== 実行 ====================
data = load_data(DATA_DIR, FILES)
if GROUP_BY not in data.columns:
    raise ValueError(f"GROUP_BY='{GROUP_BY}' がデータに見つかりません。columns: {list(data.columns)[:10]} ...")

groups = data[GROUP_BY].fillna("unknown").unique().tolist()
print("Groups:", groups)

all_res = []
out_dir = Path("model_outputs")
for g in groups:
    sub = data[data[GROUP_BY]==g].copy()
    res_g = run_group(sub, str(g), out_dir)
    all_res.append(res_g)

all_res = pd.concat(all_res, axis=0, ignore_index=True)
all_res.to_csv(out_dir / "occupied_models_all_groups.csv", index=False)

# 集計（macro平均）
agg = (all_res.groupby("model")[["accuracy","precision","recall","f1","auc"]]
       .mean(numeric_only=True)
       .reset_index())
agg.to_csv(out_dir / "occupied_models_group_macro_avg.csv", index=False)
display(agg)
print("Saved per-group & aggregate CSVs to:", out_dir)
```

```python
# ==================== 設定 ====================
DATA_DIR = "."  # CSVのあるパス（例: "/content/drive/MyDrive/sensor_data"）
FILES = [
    "living_kitchen0916_with_activity_full.csv",
    "sleepingroom0916_.csv",
    "washitsu0915_with_activity_full.csv",
]
GROUP_BY = "source_file"   # "source_file" か "Place"
RUN_NN = False             # Trueで LSTM/TCN も実行（TensorFlowが必要）
WINDOW = 60                # LSTM/TCN の窓長
COVERAGE_MIN = 0.30        # 学習(train)内の非欠損率しきい値（列除去）
CORR_DROP_TH = 0.98        # |corr(feature, Occupied)| > TH の列を除外（リーク防止）
DEV_RATIO = 0.20           # 学習末尾の dev 割合（閾値最適化に使用）
MAX_ROWS_PER_GROUP = 50000 # 各部屋あたり上限行数
NN_EPOCHS = 8
NN_BATCH = 128

# センサ判定のための名前パターン（小文字で比較）
PATTERNS = {
    "co2":      ["co2", "co₂"],
    "temp":     ["temp", "temperature"],
    "humidity": ["humid"],
    "lux":      ["lux", "illum", "light"],
    "pir":      ["pir", "motion"],
    "door":     ["door", "open", "close"],
    "wifi":     ["wifi", "wi-fi", "sniff", "mac", "rssi", "client"],
    "appliance":["power", "watt", "plug", "onoff", "ac", "tv", "microwave", "rice", "heater", "fan"],
}
LEAKY_NAME_PATTERNS = ["occupied", "presence", "activity", "label"]  # ラベルや派生名は除外

# ==================== 共通ユーティリティ ====================
import os, numpy as np, pandas as pd
from pathlib import Path

def load_data(base_dir, files):
    dfs = []
    for name in files:
        fp = Path(base_dir) / name
        if not fp.exists():
            raise FileNotFoundError(f"Missing file: {fp}")
        df = pd.read_csv(fp, low_memory=False)
        if "timestamp" in df.columns:
            try:
                df["timestamp"] = pd.to_datetime(df["timestamp"])
            except Exception:
                pass
        df["source_file"] = name
        dfs.append(df)
    data = pd.concat(dfs, axis=0, ignore_index=True, sort=False)
    for col in ["Occupied", "Activity", "Place"]:
        if col not in data.columns:
            data[col] = np.nan
    if "timestamp" in data.columns:
        data = data.sort_values("timestamp").reset_index(drop=True)
    return data

from sklearn.model_selection import train_test_split
def time_split(X, y, test_size=0.2):
    # 時系列（shuffle=False）
    Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=test_size, shuffle=False)
    # 片側単一クラスなら層別シャッフルへフォールバック
    if len(pd.Series(ytr).dropna().unique()) < 2 or len(pd.Series(yte).dropna().unique()) < 2:
        mask = y.notna()
        Xv, yv = X[mask], y[mask]
        try:
            Xtr, Xte, ytr, yte = train_test_split(Xv, yv, test_size=test_size, shuffle=True, stratify=yv)
        except Exception:
            pass
    return Xtr, Xte, ytr, yte

from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score, f1_score
def eval_bin(y_true, y_pred, y_prob=None):
    y_true = pd.Series(y_true).astype(int)
    y_pred = pd.Series(y_pred).astype(int)
    acc = accuracy_score(y_true, y_pred)
    pre, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    out = {"accuracy": float(acc), "precision": float(pre), "recall": float(rec), "f1": float(f1)}
    if y_prob is not None and len(np.unique(y_true)) == 2:
        try:
            out["auc"] = float(roc_auc_score(y_true, y_prob))
        except Exception:
            pass
    return out

def find_best_threshold(y_true, y_prob, grid=np.linspace(0.1, 0.9, 17)):
    y_true = pd.Series(y_true).astype(int)
    best_thr, best_f1 = 0.5, -1.0
    for thr in grid:
        f1 = f1_score(y_true, (y_prob >= thr).astype(int), zero_division=0)
        if f1 > best_f1:
            best_f1, best_thr = f1, thr
    return float(best_thr), float(best_f1)

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
def build_preprocess(X):
    cat_cols = [c for c in X.columns if X[c].dtype == "object"]
    num_cols = [c for c in X.columns if c not in cat_cols]
    num_pipe = Pipeline([("imputer", SimpleImputer(strategy="median")),
                         ("scaler", StandardScaler())])
    cat_pipe = Pipeline([("imputer", SimpleImputer(strategy="most_frequent")),
                         ("onehot", OneHotEncoder(handle_unknown="ignore"))])
    return ColumnTransformer([("num", num_pipe, num_cols),
                              ("cat", cat_pipe, cat_cols)])

def select_by_coverage(Xtr, min_ratio=COVERAGE_MIN):
    cov = Xtr.notna().mean()
    keep = cov[cov >= min_ratio].index.tolist()
    return keep

def drop_leaky_by_name(cols):
    keep = []
    for c in cols:
        lc = c.lower()
        if any(p in lc for p in LEAKY_NAME_PATTERNS):
            continue
        keep.append(c)
    return keep

def drop_leaky_by_corr(Xtr_df, ytr_ser, cols, th=CORR_DROP_TH):
    num_cols = [c for c in cols if pd.api.types.is_numeric_dtype(Xtr_df[c])]
    if len(num_cols) == 0:
        return cols
    corr = pd.concat([Xtr_df[num_cols], ytr_ser.reset_index(drop=True)], axis=1).corr(numeric_only=True)
    ycorr = corr[ytr_ser.name].drop(labels=[ytr_ser.name], errors="ignore")
    to_drop = set(ycorr[ycorr.abs() > th].index.tolist())
    return [c for c in cols if c not in to_drop]

# センサごとの列を推定
def infer_sensor_buckets(columns):
    buckets = {"co2":[], "temp":[], "humidity":[], "lux":[], "pir":[], "door":[], "wifi":[], "appliance":[], "others":[]}
    for c in columns:
        cl = c.lower()
        if any(p in cl for p in PATTERNS["co2"]): buckets["co2"].append(c); continue
        if any(p in cl for p in PATTERNS["temp"]): buckets["temp"].append(c); continue
        if any(p in cl for p in PATTERNS["humidity"]): buckets["humidity"].append(c); continue
        if any(p in cl for p in PATTERNS["lux"]): buckets["lux"].append(c); continue
        if any(p in cl for p in PATTERNS["pir"]): buckets["pir"].append(c); continue
        if any(p in cl for p in PATTERNS["door"]): buckets["door"].append(c); continue
        if any(p in cl for p in PATTERNS["wifi"]): buckets["wifi"].append(c); continue
        if any(p in cl for p in PATTERNS["appliance"]): buckets["appliance"].append(c); continue
        buckets["others"].append(c)
    return buckets

# グループ（複数センサ）定義
SENSOR_GROUPS = {
    "env_all": ["co2", "temp", "humidity", "lux"],       # 環境系
    "motion_all": ["pir", "door", "wifi"],                # 人の動き/存在に直結
    "appliance": ["appliance"],                           # 家電状態
    "env_plus_pir": ["co2", "temp", "humidity", "lux", "pir"],  # 例: 環境+PIR
}

# NN 用（オプション）
def build_sequences(X_arr, y_ser, window=WINDOW, step=1):
    Xs, ys = [], []
    for i in range(window, len(X_arr), step):
        Xs.append(X_arr[i-window:i, :])
        ys.append(int(y_ser.iloc[i]))
    return np.array(Xs), np.array(ys)

def cap_samples(Xs, ys, mx=8000):
    if len(Xs) > mx:
        idx = np.linspace(0, len(Xs)-1, mx).astype(int)
        return Xs[idx], ys[idx]
    return Xs, ys

from sklearn.utils.class_weight import compute_class_weight
def compute_cw(y):
    y = pd.Series(y).astype(int)
    classes = np.sort(y.unique())
    if len(classes) < 2:
        return None
    cw = compute_class_weight(class_weight="balanced", classes=classes, y=y)
    return {int(c): float(w) for c, w in zip(classes, cw)}

from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from scipy.special import expit

def run_once(df_group, group_name, feat_cols, tag, out_dir):
    """ある部屋グループに対して、指定された feat_cols のみで RF/HGB(+NN) を学習・評価"""
    # ラベル整備
    df = df_group[df_group["Occupied"].notna()].copy()
    if len(df) < 200:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"too few rows"}])
    df["Occupied"] = df["Occupied"].astype(int)
    if len(df["Occupied"].unique()) < 2:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"only one class"}])

    # 行数制限
    if len(df) > MAX_ROWS_PER_GROUP:
        idx = np.linspace(0, len(df)-1, MAX_ROWS_PER_GROUP).astype(int)
        df = df.iloc[idx].reset_index(drop=True)

    # 特徴量の起点
    X = df[feat_cols].copy()
    y = df["Occupied"].astype(int)

    # 分割
    Xtr, Xte, ytr, yte = time_split(X, y, 0.2)
    if len(pd.Series(ytr).unique()) < 2 or len(pd.Series(yte).unique()) < 2:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"single-class after split"}])

    # 低カバレッジ→リーク除去（名前/相関）
    keep = select_by_coverage(Xtr, COVERAGE_MIN)
    keep = [c for c in keep if c in feat_cols]  # feat_cols の中で
    keep = drop_leaky_by_name(keep)
    keep = drop_leaky_by_corr(Xtr[keep], ytr, keep, th=CORR_DROP_TH)
    if len(keep) == 0:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"no columns after filters"}])
    Xtr, Xte = Xtr[keep], Xte[keep]

    # 前処理
    preprocess = build_preprocess(Xtr)

    # dev（閾値最適化用）
    dev_n = max(int(len(Xtr)*DEV_RATIO), 100)
    if len(Xtr) <= dev_n + 100:
        dev_n = max(50, int(len(Xtr)*0.1))
    Xtr_tr, Xtr_dev = Xtr.iloc[:-dev_n], Xtr.iloc[-dev_n:]
    ytr_tr, ytr_dev = ytr.iloc[:-dev_n], ytr.iloc[-dev_n:]

    cw = compute_cw(ytr_tr)  # None 可
    results = []

    # ========== RandomForest ==========
    rf = Pipeline([
        ("prep", preprocess),
        ("clf", RandomForestClassifier(
            n_estimators=240,
            class_weight=("balanced" if cw is not None else None),
            random_state=42, n_jobs=-1
        ))
    ])
    rf.fit(Xtr_tr, ytr_tr)
    rf_prob_dev = rf.predict_proba(Xtr_dev)[:, 1]
    thr_rf, _ = find_best_threshold(ytr_dev, rf_prob_dev)
    rf_prob = rf.predict_proba(Xte)[:, 1]
    rf_pred = (rf_prob >= thr_rf).astype(int)
    results.append({"group": group_name, "feature_set": tag, "model":"RandomForest", **eval_bin(yte, rf_pred, rf_prob)})

    # ========== HGB (XGB相当) ==========
    preprocess.fit(Xtr_tr)
    Xtr_tr_mat = preprocess.transform(Xtr_tr); Xtr_dev_mat = preprocess.transform(Xtr_dev); Xte_mat = preprocess.transform(Xte)
    if hasattr(Xtr_tr_mat, "toarray"):
        Xtr_tr_mat, Xtr_dev_mat, Xte_mat = Xtr_tr_mat.toarray(), Xtr_dev_mat.toarray(), Xte_mat.toarray()

    hgb = HistGradientBoostingClassifier(learning_rate=0.1, random_state=42)
    if cw is not None:
        w_tr = pd.Series(ytr_tr).map(lambda v: cw.get(int(v), 1.0)).values
        hgb.fit(Xtr_tr_mat, ytr_tr, sample_weight=w_tr)
    else:
        hgb.fit(Xtr_tr_mat, ytr_tr)

    hgb_prob_dev = expit(hgb.decision_function(Xtr_dev_mat))
    thr_hgb, _ = find_best_threshold(ytr_dev, hgb_prob_dev)
    hgb_prob = expit(hgb.decision_function(Xte_mat))
    hgb_pred = (hgb_prob >= thr_hgb).astype(int)
    results.append({"group": group_name, "feature_set": tag, "model":"SklearnHGB(XGB-like)", **eval_bin(yte, hgb_pred, hgb_prob)})

    # ========== NN (任意) ==========
    if RUN_NN:
        try:
            import tensorflow as tf
            from tensorflow import keras
            from tensorflow.keras import layers
            from tensorflow.keras.callbacks import EarlyStopping

            # train-fit 済み preprocess で全体変換
            X_all_mat = preprocess.transform(df[keep])
            if hasattr(X_all_mat, "toarray"):
                X_all_mat = X_all_mat.toarray()

            train_end = len(Xtr)
            # 学習 seq / テスト seq
            X_tr_all, y_tr_all = build_sequences(X_all_mat[:train_end], y.iloc[:train_end], window=WINDOW, step=1)
            X_te_seq, y_te_seq = build_sequences(X_all_mat[train_end:], y.iloc[train_end:], window=WINDOW, step=1)

            # dev（系列）: train の 20%
            dev_len = max(int(len(X_tr_all)*DEV_RATIO), 200)
            if len(X_tr_all) <= dev_len + 100:
                dev_len = max(50, int(len(X_tr_all)*0.1))
            Xtr_seq_tr, Xtr_seq_dev = X_tr_all[:-dev_len], X_tr_all[-dev_len:]
            ytr_seq_tr, ytr_seq_dev = y_tr_all[:-dev_len], y_tr_all[-dev_len:]

            Xtr_seq_tr, ytr_seq_tr = cap_samples(Xtr_seq_tr, ytr_seq_tr, 8000)
            Xtr_seq_dev, ytr_seq_dev = cap_samples(Xtr_seq_dev, ytr_seq_dev, 3000)
            X_te_seq, y_te_seq     = cap_samples(X_te_seq, y_te_seq, 8000)

            ok = len(Xtr_seq_tr)>100 and len(np.unique(ytr_seq_tr))==2 and len(np.unique(y_te_seq))==2

            if ok:
                input_shape = (Xtr_seq_tr.shape[1], Xtr_seq_tr.shape[2])
                cw_nn = compute_cw(pd.Series(ytr_seq_tr))
                cbs = [EarlyStopping(monitor="val_loss", patience=2, restore_best_weights=True)]

                # --- LSTM ---
                lstm = keras.Sequential([
                    layers.Input(shape=input_shape),
                    layers.Masking(mask_value=0.0),
                    layers.LSTM(64),
                    layers.Dropout(0.2),
                    layers.Dense(32, activation="relu"),
                    layers.Dense(1, activation="sigmoid")
                ])
                lstm.compile(optimizer=keras.optimizers.Adam(1e-3), loss="binary_crossentropy")
                lstm.fit(Xtr_seq_tr, ytr_seq_tr, epochs=NN_EPOCHS, batch_size=NN_BATCH,
                         validation_data=(Xtr_seq_dev, ytr_seq_dev),
                         class_weight=cw_nn, callbacks=cbs, verbose=0)
                y_prob_dev = lstm.predict(Xtr_seq_dev, verbose=0).ravel()
                thr_lstm, _ = find_best_threshold(ytr_seq_dev, y_prob_dev)
                y_prob = lstm.predict(X_te_seq, verbose=0).ravel()
                y_pred = (y_prob >= thr_lstm).astype(int)
                results.append({"group": group_name, "feature_set": tag, "model":"LSTM", **eval_bin(y_te_seq, y_pred, y_prob)})

                # --- TCN ---
                inp = layers.Input(shape=input_shape)
                x = inp
                for d in [1,2,4,8]:
                    x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
                    x = layers.ReLU()(x)
                    x = layers.Dropout(0.2)(x)
                x = layers.GlobalAveragePooling1D()(x)
                x = layers.Dense(32, activation="relu")(x)
                out = layers.Dense(1, activation="sigmoid")(x)
                tcn = keras.Model(inp, out)
                tcn.compile(optimizer=keras.optimizers.Adam(1e-3), loss="binary_crossentropy")
                tcn.fit(Xtr_seq_tr, ytr_seq_tr, epochs=NN_EPOCHS, batch_size=NN_BATCH,
                        validation_data=(Xtr_seq_dev, ytr_seq_dev),
                        class_weight=cw_nn, callbacks=cbs, verbose=0)
                y_prob_dev = tcn.predict(Xtr_seq_dev, verbose=0).ravel()
                thr_tcn, _ = find_best_threshold(ytr_seq_dev, y_prob_dev)
                y_prob = tcn.predict(X_te_seq, verbose=0).ravel()
                y_pred = (y_prob >= thr_tcn).astype(int)
                results.append({"group": group_name, "feature_set": tag, "model":"TCN", **eval_bin(y_te_seq, y_pred, y_prob)})
        except Exception as e:
            print(f"[{group_name} / {tag}] NN skipped:", e)

    df_res = pd.DataFrame(results)
    out_dir.mkdir(parents=True, exist_ok=True)
    df_res.to_csv(out_dir / f"occupied_sensor_{tag}_{group_name}.csv", index=False)
    return df_res

# ==================== 実行（センサ単独 / グループ） ====================
out_dir = Path("model_outputs")
data = load_data(DATA_DIR, FILES)
if GROUP_BY not in data.columns:
    raise ValueError(f"GROUP_BY='{GROUP_BY}' がデータに見つかりません。columns: {list(data.columns)[:10]} ...")

groups = data[GROUP_BY].fillna("unknown").unique().tolist()
print("Groups:", groups)

all_results = []

for g in groups:
    sub = data[data[GROUP_BY]==g].copy()
    sub = sub[sub["Occupied"].notna()]
    if len(sub) < 200:
        all_results.append(pd.DataFrame([{"group": g, "feature_set":"N/A", "model":"N/A", "note":"too few rows"}]))
        continue

    # 特徴候補（学習に使わない列を除く）
    exclude = ["timestamp", "Place", "Activity", "Occupied", "Number of People", "source_file"]
    cand_cols = [c for c in sub.columns if c not in exclude]

    # センサバケツ推定
    buckets = infer_sensor_buckets(cand_cols)

    # -------- センサ単独 --------
    for sensor_key, cols in buckets.items():
        if len(cols) == 0:
            continue
        res = run_once(sub, str(g), cols, tag=f"single_{sensor_key}", out_dir=out_dir)
        all_results.append(res)

    # -------- グループ別 --------
    for grp_name, keys in SENSOR_GROUPS.items():
        cols = []
        for k in keys:
            cols += buckets.get(k, [])
        cols = sorted(list(set(cols)))
        if len(cols) == 0:
            continue
        res = run_once(sub, str(g), cols, tag=f"group_{grp_name}", out_dir=out_dir)
        all_results.append(res)

# 結合と集計
all_results = pd.concat(all_results, axis=0, ignore_index=True)
all_results.to_csv(out_dir / "occupied_sensor_ablation_all.csv", index=False)

# サマリ（モデル×feature_set の macro平均）
summary = (all_results
           .dropna(subset=["accuracy"])
           .groupby(["feature_set","model"])[["accuracy","precision","recall","f1","auc"]]
           .mean(numeric_only=True)
           .reset_index()
           .sort_values(["feature_set","model"]))
summary.to_csv(out_dir / "occupied_sensor_ablation_summary.csv", index=False)
display(summary)
print("Saved results to:", out_dir)
```

```python
# ==================== 設定 ====================
DATA_DIR = "."  # CSVのあるパス（例: "/content/drive/MyDrive/sensor_data"）
FILES = [
    "living_kitchen0916_with_activity_full.csv",
    "sleepingroom0916_.csv",
    "washitsu0915_with_activity_full.csv",
]
GROUP_BY = "source_file"   # "source_file" か "Place"
RUN_NN = True              # Trueで LSTM/TCN も実行（TensorFlowが必要）
WINDOW = 60                # LSTM/TCN の系列窓長
COVERAGE_MIN = 0.30        # 学習(train)内の非欠損率しきい値（列除去）
CORR_DROP_TH = 0.98        # |corr(feature, Occupied)| > TH の列を除外（リーク防止）
DEV_RATIO = 0.20           # 学習末尾の dev 割合（閾値最適化に使用）
MAX_ROWS_PER_GROUP = 50000 # 各部屋あたり上限行数
NN_EPOCHS = 8
NN_BATCH = 128

# センサ判定のための名前パターン（小文字で比較）— PIR等を強化
PATTERNS = {
    "co2":      ["co2","co₂"],
    "temp":     ["temp","temperature"],
    "humidity": ["humid","humidity"],
    "lux":      ["lux","illum","light"],
    "pir":      ["pir","motion","pir_http_","pir-","pir_","pir("],
    "door":     ["door","open","close"],
    "wifi":     ["wifi","wi-fi","sniff","rssi","client","sta_","mac_"],
    "appliance":["power","watt","plug","onoff","ac","tv","microwave","rice","heater","fan"],
}
LEAKY_NAME_PATTERNS = ["occupied","presence","activity","label"]

# グループ定義（必要に応じて編集可）
SENSOR_GROUPS = {
    "env_all": ["co2", "temp", "humidity", "lux"],          # 環境系
    "motion_all": ["pir", "door", "wifi"],                   # 存在直結
    "appliance": ["appliance"],                              # 家電状態
    "env_plus_pir": ["co2", "temp", "humidity", "lux", "pir"]# 例: 環境+PIR
}

# ==================== 共通ユーティリティ ====================
import os, numpy as np, pandas as pd
from pathlib import Path

def load_data(base_dir, files):
    dfs = []
    for name in files:
        fp = Path(base_dir) / name
        if not fp.exists():
            raise FileNotFoundError(f"Missing file: {fp}")
        df = pd.read_csv(fp, low_memory=False)
        if "timestamp" in df.columns:
            try:
                df["timestamp"] = pd.to_datetime(df["timestamp"])
            except Exception:
                pass
        df["source_file"] = name
        dfs.append(df)
    data = pd.concat(dfs, axis=0, ignore_index=True, sort=False)
    for col in ["Occupied", "Activity", "Place"]:
        if col not in data.columns:
            data[col] = np.nan
    if "timestamp" in data.columns:
        data = data.sort_values("timestamp").reset_index(drop=True)
    return data

from sklearn.model_selection import train_test_split
def time_split(X, y, test_size=0.2):
    Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=test_size, shuffle=False)
    if len(pd.Series(ytr).dropna().unique()) < 2 or len(pd.Series(yte).dropna().unique()) < 2:
        mask = y.notna()
        Xv, yv = X[mask], y[mask]
        try:
            Xtr, Xte, ytr, yte = train_test_split(Xv, yv, test_size=test_size, shuffle=True, stratify=yv)
        except Exception:
            pass
    return Xtr, Xte, ytr, yte

from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score, f1_score, average_precision_score, confusion_matrix
def eval_bin(y_true, y_pred, y_prob=None):
    y_true = pd.Series(y_true).astype(int)
    y_pred = pd.Series(y_pred).astype(int)
    acc = accuracy_score(y_true, y_pred)
    pre, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    out = {"accuracy": float(acc), "precision": float(pre), "recall": float(rec), "f1": float(f1)}
    if y_prob is not None and len(np.unique(y_true)) == 2:
        try:
            out["auc"] = float(roc_auc_score(y_true, y_prob))
        except Exception:
            pass
        try:
            out["pr_auc"] = float(average_precision_score(y_true, y_prob))
        except Exception:
            pass
    # 混同行列（参考：列は出力CSVに含めない）
    try:
        tn, fp, fn, tp = confusion_matrix(y_true, y_pred, labels=[0,1]).ravel()
        out.update({"tn":int(tn), "fp":int(fp), "fn":int(fn), "tp":int(tp)})
    except Exception:
        pass
    return out

def find_best_threshold(y_true, y_prob, grid=np.linspace(0.1, 0.9, 17)):
    y_true = pd.Series(y_true).astype(int)
    best_thr, best_f1 = 0.5, -1.0
    for thr in grid:
        f1 = f1_score(y_true, (y_prob >= thr).astype(int), zero_division=0)
        if f1 > best_f1:
            best_f1, best_thr = f1, thr
    return float(best_thr), float(best_f1)

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
def build_preprocess(X):
    cat_cols = [c for c in X.columns if X[c].dtype == "object"]
    num_cols = [c for c in X.columns if c not in cat_cols]
    num_pipe = Pipeline([("imputer", SimpleImputer(strategy="median")),
                         ("scaler", StandardScaler())])
    cat_pipe = Pipeline([("imputer", SimpleImputer(strategy="most_frequent")),
                         ("onehot", OneHotEncoder(handle_unknown="ignore"))])
    return ColumnTransformer([("num", num_pipe, num_cols),
                              ("cat", cat_pipe, cat_cols)])

def select_by_coverage(Xtr, min_ratio=COVERAGE_MIN):
    cov = Xtr.notna().mean()
    keep = cov[cov >= min_ratio].index.tolist()
    return keep

def drop_leaky_by_name(cols):
    keep = []
    for c in cols:
        lc = c.lower()
        if any(p in lc for p in LEAKY_NAME_PATTERNS):
            continue
        keep.append(c)
    return keep

def drop_leaky_by_corr(Xtr_df, ytr_ser, cols, th=CORR_DROP_TH):
    num_cols = [c for c in cols if pd.api.types.is_numeric_dtype(Xtr_df[c])]
    if len(num_cols) == 0:
        return cols
    corr = pd.concat([Xtr_df[num_cols], ytr_ser.reset_index(drop=True)], axis=1).corr(numeric_only=True)
    ycorr = corr[ytr_ser.name].drop(labels=[ytr_ser.name], errors="ignore")
    to_drop = set(ycorr[ycorr.abs() > th].index.tolist())
    return [c for c in cols if c not in to_drop]

def infer_sensor_buckets(columns):
    buckets = {"co2":[], "temp":[], "humidity":[], "lux":[], "pir":[], "door":[], "wifi":[], "appliance":[], "others":[]}
    for c in columns:
        cl = c.lower()
        if any(p in cl for p in PATTERNS["co2"]): buckets["co2"].append(c); continue
        if any(p in cl for p in PATTERNS["temp"]): buckets["temp"].append(c); continue
        if any(p in cl for p in PATTERNS["humidity"]): buckets["humidity"].append(c); continue
        if any(p in cl for p in PATTERNS["lux"]): buckets["lux"].append(c); continue
        if any(p in cl for p in PATTERNS["pir"]): buckets["pir"].append(c); continue
        if any(p in cl for p in PATTERNS["door"]): buckets["door"].append(c); continue
        if any(p in cl for p in PATTERNS["wifi"]): buckets["wifi"].append(c); continue
        if any(p in cl for p in PATTERNS["appliance"]): buckets["appliance"].append(c); continue
        buckets["others"].append(c)
    return buckets

# NN 用
def build_sequences(X_arr, y_ser, window=WINDOW, step=1):
    Xs, ys = [], []
    for i in range(window, len(X_arr), step):
        Xs.append(X_arr[i-window:i, :])
        ys.append(int(y_ser.iloc[i]))
    return np.array(Xs), np.array(ys)

def cap_samples(Xs, ys, mx=8000):
    if len(Xs) > mx:
        idx = np.linspace(0, len(Xs)-1, mx).astype(int)
        return Xs[idx], ys[idx]
    return Xs, ys

from sklearn.utils.class_weight import compute_class_weight
def compute_cw(y):
    y = pd.Series(y).astype(int)
    classes = np.sort(y.unique())
    if len(classes) < 2:
        return None
    cw = compute_class_weight(class_weight="balanced", classes=classes, y=y)
    return {int(c): float(w) for c, w in zip(classes, cw)}

# モデル
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from scipy.special import expit

def run_once(df_group, group_name, feat_cols, tag, out_dir):
    """ある部屋グループに対して、指定された feat_cols のみで RF/HGB/LSTM/TCN を学習・評価"""
    # ラベル整備
    df = df_group[df_group["Occupied"].notna()].copy()
    if len(df) < 200:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"too few rows"}])
    df["Occupied"] = df["Occupied"].astype(int)
    if len(df["Occupied"].unique()) < 2:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"only one class"}])

    # 行数制限
    if len(df) > MAX_ROWS_PER_GROUP:
        idx = np.linspace(0, len(df)-1, MAX_ROWS_PER_GROUP).astype(int)
        df = df.iloc[idx].reset_index(drop=True)

    # 特徴候補
    X = df[feat_cols].copy()
    y = df["Occupied"].astype(int)

    # 分割
    Xtr, Xte, ytr, yte = time_split(X, y, 0.2)
    if len(pd.Series(ytr).unique()) < 2 or len(pd.Series(yte).unique()) < 2:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"single-class after split"}])

    # 低カバレッジ→リーク除去（名前/相関）
    keep = select_by_coverage(Xtr, COVERAGE_MIN)
    keep = [c for c in keep if c in feat_cols]
    keep = drop_leaky_by_name(keep)
    keep = drop_leaky_by_corr(Xtr[keep], ytr, keep, th=CORR_DROP_TH)
    if len(keep) == 0:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"no columns after filters"}])

    Xtr, Xte = Xtr[keep], Xte[keep]

    # 前処理
    preprocess = build_preprocess(Xtr)

    # dev（閾値最適化用）
    dev_n = max(int(len(Xtr)*DEV_RATIO), 100)
    if len(Xtr) <= dev_n + 100:
        dev_n = max(50, int(len(Xtr)*0.1))
    Xtr_tr, Xtr_dev = Xtr.iloc[:-dev_n], Xtr.iloc[-dev_n:]
    ytr_tr, ytr_dev = ytr.iloc[:-dev_n], ytr.iloc[-dev_n:]

    cw = compute_cw(ytr_tr)  # None 可
    results = []

    # ========== RandomForest ==========
    rf = Pipeline([
        ("prep", preprocess),
        ("clf", RandomForestClassifier(
            n_estimators=240,
            class_weight=("balanced" if cw is not None else None),
            random_state=42, n_jobs=-1
        ))
    ])
    rf.fit(Xtr_tr, ytr_tr)
    rf_prob_dev = rf.predict_proba(Xtr_dev)[:, 1]
    thr_rf, _ = find_best_threshold(ytr_dev, rf_prob_dev)
    rf_prob = rf.predict_proba(Xte)[:, 1]
    rf_pred = (rf_prob >= thr_rf).astype(int)
    results.append({"group": group_name, "feature_set": tag, "model":"RandomForest", **eval_bin(yte, rf_pred, rf_prob)})

    # ========== HGB (XGB相当) ==========
    preprocess.fit(Xtr_tr)
    Xtr_tr_mat = preprocess.transform(Xtr_tr); Xtr_dev_mat = preprocess.transform(Xtr_dev); Xte_mat = preprocess.transform(Xte)
    if hasattr(Xtr_tr_mat, "toarray"):
        Xtr_tr_mat, Xtr_dev_mat, Xte_mat = Xtr_tr_mat.toarray(), Xtr_dev_mat.toarray(), Xte_mat.toarray()

    hgb = HistGradientBoostingClassifier(learning_rate=0.1, random_state=42)
    if cw is not None:
        w_tr = pd.Series(ytr_tr).map(lambda v: cw.get(int(v), 1.0)).values
        hgb.fit(Xtr_tr_mat, ytr_tr, sample_weight=w_tr)
    else:
        hgb.fit(Xtr_tr_mat, ytr_tr)

    hgb_prob_dev = expit(hgb.decision_function(Xtr_dev_mat))
    thr_hgb, _ = find_best_threshold(ytr_dev, hgb_prob_dev)
    hgb_prob = expit(hgb.decision_function(Xte_mat))
    hgb_pred = (hgb_prob >= thr_hgb).astype(int)
    results.append({"group": group_name, "feature_set": tag, "model":"SklearnHGB(XGB-like)", **eval_bin(yte, hgb_pred, hgb_prob)})

    # ========== NN（LSTM / TCN） ==========
    if RUN_NN:
        try:
            import tensorflow as tf
            from tensorflow import keras
            from tensorflow.keras import layers
            from tensorflow.keras.callbacks import EarlyStopping

            # train-fit 済み preprocess で全体変換
            X_all_mat = preprocess.transform(df[keep])
            if hasattr(X_all_mat, "toarray"):
                X_all_mat = X_all_mat.toarray()

            # シーケンス構築（train: Xtr の長さまで、test: その後）
            train_end = len(Xtr)
            X_tr_all, y_tr_all = build_sequences(X_all_mat[:train_end], y.iloc[:train_end], window=WINDOW, step=1)
            X_te_seq, y_te_seq = build_sequences(X_all_mat[train_end:], y.iloc[train_end:], window=WINDOW, step=1)

            # dev: train の末尾 DEV_RATIO
            dev_len = max(int(len(X_tr_all)*DEV_RATIO), 200)
            if len(X_tr_all) <= dev_len + 100:
                dev_len = max(50, int(len(X_tr_all)*0.1))
            Xtr_seq_tr, Xtr_seq_dev = X_tr_all[:-dev_len], X_tr_all[-dev_len:]
            ytr_seq_tr, ytr_seq_dev = y_tr_all[:-dev_len], y_tr_all[-dev_len:]

            # サンプル上限
            Xtr_seq_tr, ytr_seq_tr = cap_samples(Xtr_seq_tr, ytr_seq_tr, 8000)
            Xtr_seq_dev, ytr_seq_dev = cap_samples(Xtr_seq_dev, ytr_seq_dev, 3000)
            X_te_seq, y_te_seq     = cap_samples(X_te_seq, y_te_seq, 8000)

            ok = len(Xtr_seq_tr)>100 and len(np.unique(ytr_seq_tr))==2 and len(np.unique(y_te_seq))==2
            if ok:
                input_shape = (Xtr_seq_tr.shape[1], Xtr_seq_tr.shape[2])
                cw_nn = compute_cw(pd.Series(ytr_seq_tr))
                cbs = [EarlyStopping(monitor="val_loss", patience=2, restore_best_weights=True)]

                # --- LSTM ---
                lstm = keras.Sequential([
                    layers.Input(shape=input_shape),
                    layers.Masking(mask_value=0.0),
                    layers.LSTM(64),
                    layers.Dropout(0.2),
                    layers.Dense(32, activation="relu"),
                    layers.Dense(1, activation="sigmoid")
                ])
                lstm.compile(optimizer=keras.optimizers.Adam(1e-3), loss="binary_crossentropy")
                lstm.fit(Xtr_seq_tr, ytr_seq_tr, epochs=NN_EPOCHS, batch_size=NN_BATCH,
                         validation_data=(Xtr_seq_dev, ytr_seq_dev),
                         class_weight=cw_nn, callbacks=cbs, verbose=0)
                y_prob_dev = lstm.predict(Xtr_seq_dev, verbose=0).ravel()
                thr_lstm, _ = find_best_threshold(ytr_seq_dev, y_prob_dev)
                y_prob = lstm.predict(X_te_seq, verbose=0).ravel()
                y_pred = (y_prob >= thr_lstm).astype(int)
                results.append({"group": group_name, "feature_set": tag, "model":"LSTM", **eval_bin(y_te_seq, y_pred, y_prob)})

                # --- TCN（因果 Dilated Conv）---
                inp = layers.Input(shape=input_shape)
                x = inp
                for d in [1,2,4,8]:
                    x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
                    x = layers.ReLU()(x)
                    x = layers.Dropout(0.2)(x)
                x = layers.GlobalAveragePooling1D()(x)
                x = layers.Dense(32, activation="relu")(x)
                out = layers.Dense(1, activation="sigmoid")(x)
                tcn = keras.Model(inp, out)
                tcn.compile(optimizer=keras.optimizers.Adam(1e-3), loss="binary_crossentropy")
                tcn.fit(Xtr_seq_tr, ytr_seq_tr, epochs=NN_EPOCHS, batch_size=NN_BATCH,
                        validation_data=(Xtr_seq_dev, ytr_seq_dev),
                        class_weight=cw_nn, callbacks=cbs, verbose=0)
                y_prob_dev = tcn.predict(Xtr_seq_dev, verbose=0).ravel()
                thr_tcn, _ = find_best_threshold(ytr_seq_dev, y_prob_dev)
                y_prob = tcn.predict(X_te_seq, verbose=0).ravel()
                y_pred = (y_prob >= thr_tcn).astype(int)
                results.append({"group": group_name, "feature_set": tag, "model":"TCN", **eval_bin(y_te_seq, y_pred, y_prob)})
        except Exception as e:
            print(f"[{group_name} / {tag}] NN skipped:", e)

    df_res = pd.DataFrame(results)
    out_dir.mkdir(parents=True, exist_ok=True)
    df_res.to_csv(out_dir / f"occupied_sensor_{tag}_{group_name}.csv", index=False)
    return df_res

# ==================== 実行（センサ単独 / グループ × 4モデル） ====================
out_dir = Path("model_outputs")
data = load_data(DATA_DIR, FILES)
if GROUP_BY not in data.columns:
    raise ValueError(f"GROUP_BY='{GROUP_BY}' がデータに見つかりません。columns: {list(data.columns)[:10]} ...")

groups = data[GROUP_BY].fillna("unknown").unique().tolist()
print("Groups:", groups)

all_results = []

for g in groups:
    sub = data[data[GROUP_BY]==g].copy()
    sub = sub[sub["Occupied"].notna()]
    if len(sub) < 200:
        all_results.append(pd.DataFrame([{"group": g, "feature_set":"N/A", "model":"N/A", "note":"too few rows"}]))
        continue

    exclude = ["timestamp", "Place", "Activity", "Occupied", "Number of People", "source_file"]
    cand_cols = [c for c in sub.columns if c not in exclude]

    buckets = infer_sensor_buckets(cand_cols)

    # センサ単独
    for sensor_key, cols in buckets.items():
        if len(cols) == 0:
            continue
        res = run_once(sub, str(g), cols, tag=f"single_{sensor_key}", out_dir=out_dir)
        all_results.append(res)

    # グループ
    for grp_name, keys in SENSOR_GROUPS.items():
        cols = []
        for k in keys:
            cols += buckets.get(k, [])
        cols = sorted(list(set(cols)))
        if len(cols) == 0:
            continue
        res = run_once(sub, str(g), cols, tag=f"group_{grp_name}", out_dir=out_dir)
        all_results.append(res)

# 結合とサマリ
all_results = pd.concat(all_results, axis=0, ignore_index=True)
all_results.to_csv(out_dir / "occupied_sensor_ablation_all_4models.csv", index=False)

summary = (all_results
           .dropna(subset=["accuracy"])
           .groupby(["feature_set","model"])[["accuracy","precision","recall","f1","auc","pr_auc"]]
           .mean(numeric_only=True)
           .reset_index()
           .sort_values(["feature_set","model"]))
summary.to_csv(out_dir / "occupied_sensor_ablation_summary_4models.csv", index=False)
display(summary)
print("Saved results to:", out_dir)
```

```python
# ==================== 設定 ====================
DATA_DIR = "."  # CSVのあるパス（例: "/content/drive/MyDrive/sensor_data"）
FILES = [
    "living_kitchen0916_with_activity_full.csv",
    "sleepingroom0916_.csv",
    "washitsu0915_with_activity_full.csv",
]
GROUP_BY = "source_file"   # "source_file" か "Place"
RUN_NN = True              # Trueで LSTM/TCN も実行（TensorFlowが必要）
USE_FOCAL = True           # Trueで Focal Loss を使用（不均衡に強い）
WINDOW = 60                # LSTM/TCN の系列窓長
COVERAGE_MIN = 0.30        # 学習(train)内の非欠損率しきい値（列除去）
CORR_DROP_TH = 0.98        # |corr(feature, Occupied)| > TH の列を除外（リーク防止）
DEV_RATIO = 0.25           # 学習末尾の dev 割合（閾値最適化に使用、従来より広げる）
MAX_ROWS_PER_GROUP = 50000 # 各部屋あたり上限行数
NN_EPOCHS = 12
NN_BATCH = 128
THRESH_GRID = np.linspace(0.05, 0.95, 19)  # 閾値最適化グリッド（細かめ）

# センサ判定のための名前パターン（小文字で比較）— PIR等を強化
PATTERNS = {
    "co2":      ["co2","co₂"],
    "temp":     ["temp","temperature"],
    "humidity": ["humid","humidity"],
    "lux":      ["lux","illum","light"],
    "pir":      ["pir","motion","pir_http_","pir-","pir_","pir("],
    "door":     ["door","open","close"],
    "wifi":     ["wifi","wi-fi","sniff","rssi","client","sta_","mac_"],
    "appliance":["power","watt","plug","onoff","ac","tv","microwave","rice","heater","fan"],
}
LEAKY_NAME_PATTERNS = ["occupied","presence","activity","label"]

# グループ定義（必要に応じて編集可）
SENSOR_GROUPS = {
    "env_all": ["co2", "temp", "humidity", "lux"],          # 環境系
    "motion_all": ["pir", "door", "wifi"],                   # 存在直結
    "appliance": ["appliance"],                              # 家電状態
    "env_plus_pir": ["co2", "temp", "humidity", "lux", "pir"]# 例: 環境+PIR
}

# ==================== 共通ユーティリティ ====================
import os, numpy as np, pandas as pd
from pathlib import Path

def load_data(base_dir, files):
    dfs = []
    for name in files:
        fp = Path(base_dir) / name
        if not fp.exists():
            raise FileNotFoundError(f"Missing file: {fp}")
        df = pd.read_csv(fp, low_memory=False)
        if "timestamp" in df.columns:
            try:
                df["timestamp"] = pd.to_datetime(df["timestamp"])
            except Exception:
                pass
        df["source_file"] = name
        dfs.append(df)
    data = pd.concat(dfs, axis=0, ignore_index=True, sort=False)
    for col in ["Occupied", "Activity", "Place"]:
        if col not in data.columns:
            data[col] = np.nan
    if "timestamp" in data.columns:
        data = data.sort_values("timestamp").reset_index(drop=True)
    return data

from sklearn.model_selection import train_test_split
def time_split(X, y, test_size=0.2):
    Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=test_size, shuffle=False)
    if len(pd.Series(ytr).dropna().unique()) < 2 or len(pd.Series(yte).dropna().unique()) < 2:
        mask = y.notna()
        Xv, yv = X[mask], y[mask]
        try:
            Xtr, Xte, ytr, yte = train_test_split(Xv, yv, test_size=test_size, shuffle=True, stratify=yv)
        except Exception:
            pass
    return Xtr, Xte, ytr, yte

from sklearn.metrics import accuracy_score, precision_recall_fscore_support, roc_auc_score, f1_score, average_precision_score, confusion_matrix
def eval_bin(y_true, y_pred, y_prob=None):
    y_true = pd.Series(y_true).astype(int)
    y_pred = pd.Series(y_pred).astype(int)
    acc = accuracy_score(y_true, y_pred)
    pre, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="binary", zero_division=0)
    out = {"accuracy": float(acc), "precision": float(pre), "recall": float(rec), "f1": float(f1)}
    if y_prob is not None and len(np.unique(y_true)) == 2:
        try:
            out["auc"] = float(roc_auc_score(y_true, y_prob))
        except Exception:
            pass
        try:
            out["pr_auc"] = float(average_precision_score(y_true, y_prob))
        except Exception:
            pass
    try:
        tn, fp, fn, tp = confusion_matrix(y_true, y_pred, labels=[0,1]).ravel()
        out.update({"tn":int(tn), "fp":int(fp), "fn":int(fn), "tp":int(tp)})
    except Exception:
        pass
    return out

def find_best_threshold(y_true, y_prob, grid=THRESH_GRID):
    y_true = pd.Series(y_true).astype(int)
    best_thr, best_f1 = 0.5, -1.0
    for thr in grid:
        f1 = f1_score(y_true, (y_prob >= thr).astype(int), zero_division=0)
        if f1 > best_f1:
            best_f1, best_thr = f1, thr
    return float(best_thr), float(best_f1)

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
def build_preprocess(X):
    cat_cols = [c for c in X.columns if X[c].dtype == "object"]
    num_cols = [c for c in X.columns if c not in cat_cols]
    num_pipe = Pipeline([("imputer", SimpleImputer(strategy="median")),
                         ("scaler", StandardScaler())])
    cat_pipe = Pipeline([("imputer", SimpleImputer(strategy="most_frequent")),
                         ("onehot", OneHotEncoder(handle_unknown="ignore"))])
    return ColumnTransformer([("num", num_pipe, num_cols),
                              ("cat", cat_pipe, cat_cols)])

def select_by_coverage(Xtr, min_ratio=COVERAGE_MIN):
    cov = Xtr.notna().mean()
    keep = cov[cov >= min_ratio].index.tolist()
    return keep

def drop_leaky_by_name(cols):
    keep = []
    for c in cols:
        lc = c.lower()
        if any(p in lc for p in LEAKY_NAME_PATTERNS):
            continue
        keep.append(c)
    return keep

def drop_leaky_by_corr(Xtr_df, ytr_ser, cols, th=CORR_DROP_TH):
    num_cols = [c for c in cols if pd.api.types.is_numeric_dtype(Xtr_df[c])]
    if len(num_cols) == 0:
        return cols
    corr = pd.concat([Xtr_df[num_cols], ytr_ser.reset_index(drop=True)], axis=1).corr(numeric_only=True)
    ycorr = corr[ytr_ser.name].drop(labels=[ytr_ser.name], errors="ignore")
    to_drop = set(ycorr[ycorr.abs() > th].index.tolist())
    return [c for c in cols if c not in to_drop]

def infer_sensor_buckets(columns):
    buckets = {"co2":[], "temp":[], "humidity":[], "lux":[], "pir":[], "door":[], "wifi":[], "appliance":[], "others":[]}
    for c in columns:
        cl = c.lower()
        if any(p in cl for p in PATTERNS["co2"]): buckets["co2"].append(c); continue
        if any(p in cl for p in PATTERNS["temp"]): buckets["temp"].append(c); continue
        if any(p in cl for p in PATTERNS["humidity"]): buckets["humidity"].append(c); continue
        if any(p in cl for p in PATTERNS["lux"]): buckets["lux"].append(c); continue
        if any(p in cl for p in PATTERNS["pir"]): buckets["pir"].append(c); continue
        if any(p in cl for p in PATTERNS["door"]): buckets["door"].append(c); continue
        if any(p in cl for p in PATTERNS["wifi"]): buckets["wifi"].append(c); continue
        if any(p in cl for p in PATTERNS["appliance"]): buckets["appliance"].append(c); continue
        buckets["others"].append(c)
    return buckets

# NN 用：系列化 + バランシング
def build_sequences(X_arr, y_ser, window=WINDOW, step=1):
    Xs, ys = [], []
    for i in range(window, len(X_arr), step):
        Xs.append(X_arr[i-window:i, :])
        ys.append(int(y_ser.iloc[i]))
    return np.array(Xs), np.array(ys)

def mk_balanced_sequences(X_arr, y_ser, window=WINDOW, step=1, max_per_class=8000):
    Xs, ys = build_sequences(X_arr, y_ser, window=window, step=step)
    if len(ys) == 0:
        return Xs, ys
    pos = np.where(ys==1)[0]; neg = np.where(ys==0)[0]
    if len(pos)==0 or len(neg)==0:
        return Xs, ys
    k = min(len(pos), len(neg), max_per_class)
    sel = np.concatenate([np.random.choice(pos, k, replace=False),
                          np.random.choice(neg, k, replace=False)])
    np.random.shuffle(sel)
    return Xs[sel], ys[sel]

def cap_samples(Xs, ys, mx=8000):
    if len(Xs) > mx:
        idx = np.linspace(0, len(Xs)-1, mx).astype(int)
        return Xs[idx], ys[idx]
    return Xs, ys

from sklearn.utils.class_weight import compute_class_weight
def compute_cw(y):
    y = pd.Series(y).astype(int)
    classes = np.sort(y.unique())
    if len(classes) < 2:
        return None
    cw = compute_class_weight(class_weight="balanced", classes=classes, y=y)
    return {int(c): float(w) for c, w in zip(classes, cw)}

# モデル
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from scipy.special import expit

# ==================== NN（LSTM/TCN）学習ヘルパ ====================
def get_callbacks():
    from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
    return [
        EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True),
        ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=2, min_lr=1e-5)
    ]

def focal_loss(gamma=2.0, alpha=0.25):
    import tensorflow as tf
    def loss(y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        eps = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, eps, 1.0-eps)
        pt = tf.where(tf.equal(y_true,1.0), y_pred, 1.0-y_pred)
        return -tf.reduce_mean(alpha * tf.pow(1.0-pt, gamma) * tf.math.log(pt))
    return loss

def build_lstm(input_shape):
    from tensorflow import keras
    from tensorflow.keras import layers
    m = keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Masking(mask_value=0.0),
        layers.LSTM(64),
        layers.Dropout(0.2),
        layers.Dense(32, activation="relu"),
        layers.Dense(1, activation="sigmoid")
    ])
    return m

def build_tcn(input_shape):
    from tensorflow import keras
    from tensorflow.keras import layers
    inp = layers.Input(shape=input_shape)
    x = inp
    for d in [1,2,4,8]:
        x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
        x = layers.ReLU()(x)
        x = layers.Dropout(0.2)(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(32, activation="relu")(x)
    out = layers.Dense(1, activation="sigmoid")(x)
    return keras.Model(inp, out)

# ==================== 1条件実行（4モデル） ====================
def run_once(df_group, group_name, feat_cols, tag, out_dir):
    """ある部屋グループに対して、指定された feat_cols のみで RF/HGB/LSTM/TCN を学習・評価"""
    # ラベル整備
    df = df_group[df_group["Occupied"].notna()].copy()
    if len(df) < 200:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"too few rows"}])
    df["Occupied"] = df["Occupied"].astype(int)
    if len(df["Occupied"].unique()) < 2:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"only one class"}])

    # 行数制限
    if len(df) > MAX_ROWS_PER_GROUP:
        idx = np.linspace(0, len(df)-1, MAX_ROWS_PER_GROUP).astype(int)
        df = df.iloc[idx].reset_index(drop=True)

    # 特徴候補
    X = df[feat_cols].copy()
    y = df["Occupied"].astype(int)

    # 分割
    Xtr, Xte, ytr, yte = time_split(X, y, 0.2)
    if len(pd.Series(ytr).unique()) < 2 or len(pd.Series(yte).unique()) < 2:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"single-class after split"}])

    # 低カバレッジ→リーク除去（名前/相関）
    keep = select_by_coverage(Xtr, COVERAGE_MIN)
    keep = [c for c in keep if c in feat_cols]
    keep = drop_leaky_by_name(keep)
    keep = drop_leaky_by_corr(Xtr[keep], ytr, keep, th=CORR_DROP_TH)
    if len(keep) == 0:
        return pd.DataFrame([{"group": group_name, "feature_set": tag, "model":"N/A", "note":"no columns after filters"}])

    Xtr, Xte = Xtr[keep], Xte[keep]

    # 前処理
    preprocess = build_preprocess(Xtr)

    # dev（閾値最適化用）
    dev_n = max(int(len(Xtr)*DEV_RATIO), 100)
    if len(Xtr) <= dev_n + 100:
        dev_n = max(50, int(len(Xtr)*0.1))
    Xtr_tr, Xtr_dev = Xtr.iloc[:-dev_n], Xtr.iloc[-dev_n:]
    ytr_tr, ytr_dev = ytr.iloc[:-dev_n], ytr.iloc[-dev_n:]

    cw = compute_cw(ytr_tr)  # None 可
    results = []

    # ========== RandomForest ==========
    rf = Pipeline([
        ("prep", preprocess),
        ("clf", RandomForestClassifier(
            n_estimators=280,
            class_weight=("balanced" if cw is not None else None),
            random_state=42, n_jobs=-1
        ))
    ])
    rf.fit(Xtr_tr, ytr_tr)
    rf_prob_dev = rf.predict_proba(Xtr_dev)[:, 1]
    thr_rf, _ = find_best_threshold(ytr_dev, rf_prob_dev, grid=THRESH_GRID)
    rf_prob = rf.predict_proba(Xte)[:, 1]
    rf_pred = (rf_prob >= thr_rf).astype(int)
    results.append({"group": group_name, "feature_set": tag, "model":"RandomForest", **eval_bin(yte, rf_pred, rf_prob)})

    # ========== HGB (XGB相当) ==========
    preprocess.fit(Xtr_tr)
    Xtr_tr_mat = preprocess.transform(Xtr_tr); Xtr_dev_mat = preprocess.transform(Xtr_dev); Xte_mat = preprocess.transform(Xte)
    if hasattr(Xtr_tr_mat, "toarray"):
        Xtr_tr_mat, Xtr_dev_mat, Xte_mat = Xtr_tr_mat.toarray(), Xtr_dev_mat.toarray(), Xte_mat.toarray()

    hgb = HistGradientBoostingClassifier(learning_rate=0.1, random_state=42)
    if cw is not None:
        w_tr = pd.Series(ytr_tr).map(lambda v: cw.get(int(v), 1.0)).values
        hgb.fit(Xtr_tr_mat, ytr_tr, sample_weight=w_tr)
    else:
        hgb.fit(Xtr_tr_mat, ytr_tr)

    hgb_prob_dev = expit(hgb.decision_function(Xtr_dev_mat))
    thr_hgb, _ = find_best_threshold(ytr_dev, hgb_prob_dev, grid=THRESH_GRID)
    hgb_prob = expit(hgb.decision_function(Xte_mat))
    hgb_pred = (hgb_prob >= thr_hgb).astype(int)
    results.append({"group": group_name, "feature_set": tag, "model":"SklearnHGB(XGB-like)", **eval_bin(yte, hgb_pred, hgb_prob)})

    # ========== NN（LSTM / TCN） ==========
    if RUN_NN:
        try:
            import tensorflow as tf
            from tensorflow import keras

            # train-fit 済み preprocess で全体変換
            X_all_mat = preprocess.transform(df[keep])
            if hasattr(X_all_mat, "toarray"):
                X_all_mat = X_all_mat.toarray()

            # シーケンス（train/dev はバランス化、testはそのまま）
            train_end = len(Xtr)
            X_tr_bal, y_tr_bal = mk_balanced_sequences(X_all_mat[:train_end], y.iloc[:train_end], window=WINDOW, step=1, max_per_class=8000)
            X_dev_bal, y_dev_bal = mk_balanced_sequences(X_all_mat[:train_end], y.iloc[:train_end], window=WINDOW, step=1, max_per_class=3000)
            # devは末尾に寄せた方が時系列には良いが、ここではクラスバランスを優先
            X_te_seq,  y_te_seq  = build_sequences(X_all_mat[train_end:], y.iloc[train_end:], window=WINDOW, step=1)

            # 最低限の条件
            ok = (len(X_tr_bal)>200 and len(np.unique(y_tr_bal))==2 and len(np.unique(y_te_seq))==2)
            if ok:
                input_shape = (X_tr_bal.shape[1], X_tr_bal.shape[2])
                cbs = get_callbacks()
                loss_fn = focal_loss() if USE_FOCAL else "binary_crossentropy"

                # class_weight はバランス化済みなので基本不要（None）。必要なら下記をON
                # cw_nn = compute_cw(pd.Series(y_tr_bal))

                # --- LSTM ---
                lstm = build_lstm(input_shape)
                lstm.compile(optimizer=keras.optimizers.Adam(1e-3), loss=loss_fn)
                lstm.fit(X_tr_bal, y_tr_bal, epochs=NN_EPOCHS, batch_size=NN_BATCH,
                         validation_data=(X_dev_bal, y_dev_bal),
                         callbacks=cbs, verbose=0)
                y_prob_dev = lstm.predict(X_dev_bal, verbose=0).ravel()
                thr_lstm, _ = find_best_threshold(y_dev_bal, y_prob_dev, grid=THRESH_GRID)
                y_prob = lstm.predict(X_te_seq, verbose=0).ravel()
                y_pred = (y_prob >= thr_lstm).astype(int)
                results.append({"group": group_name, "feature_set": tag, "model":"LSTM", **eval_bin(y_te_seq, y_pred, y_prob)})

                # --- TCN ---
                tcn = build_tcn(input_shape)
                tcn.compile(optimizer=keras.optimizers.Adam(1e-3), loss=loss_fn)
                tcn.fit(X_tr_bal, y_tr_bal, epochs=NN_EPOCHS, batch_size=NN_BATCH,
                        validation_data=(X_dev_bal, y_dev_bal),
                        callbacks=cbs, verbose=0)
                y_prob_dev = tcn.predict(X_dev_bal, verbose=0).ravel()
                thr_tcn, _ = find_best_threshold(y_dev_bal, y_prob_dev, grid=THRESH_GRID)
                y_prob = tcn.predict(X_te_seq, verbose=0).ravel()
                y_pred = (y_prob >= thr_tcn).astype(int)
                results.append({"group": group_name, "feature_set": tag, "model":"TCN", **eval_bin(y_te_seq, y_pred, y_prob)})
        except Exception as e:
            print(f"[{group_name} / {tag}] NN skipped:", e)

    df_res = pd.DataFrame(results)
    out_dir.mkdir(parents=True, exist_ok=True)
    df_res.to_csv(out_dir / f"occupied_sensor_{tag}_{group_name}.csv", index=False)
    return df_res

# ==================== 実行（センサ単独 / グループ × 4モデル） ====================
out_dir = Path("model_outputs")
data = load_data(DATA_DIR, FILES)
if GROUP_BY not in data.columns:
    raise ValueError(f"GROUP_BY='{GROUP_BY}' がデータに見つかりません。columns: {list(data.columns)[:10]} ...")

groups = data[GROUP_BY].fillna("unknown").unique().tolist()
print("Groups:", groups)

all_results = []

for g in groups:
    sub = data[data[GROUP_BY]==g].copy()
    sub = sub[sub["Occupied"].notna()]
    if len(sub) < 200:
        all_results.append(pd.DataFrame([{"group": g, "feature_set":"N/A", "model":"N/A", "note":"too few rows"}]))
        continue

    exclude = ["timestamp", "Place", "Activity", "Occupied", "Number of People", "source_file"]
    cand_cols = [c for c in sub.columns if c not in exclude]

    buckets = infer_sensor_buckets(cand_cols)

    # センサ単独
    for sensor_key, cols in buckets.items():
        if len(cols) == 0:
            continue
        res = run_once(sub, str(g), cols, tag=f"single_{sensor_key}", out_dir=out_dir)
        all_results.append(res)

    # グループ
    for grp_name, keys in SENSOR_GROUPS.items():
        cols = []
        for k in keys:
            cols += buckets.get(k, [])
        cols = sorted(list(set(cols)))
        if len(cols) == 0:
            continue
        res = run_once(sub, str(g), cols, tag=f"group_{grp_name}", out_dir=out_dir)
        all_results.append(res)

# 結合とサマリ
all_results = pd.concat(all_results, axis=0, ignore_index=True)
all_results.to_csv(out_dir / "occupied_sensor_ablation_all_4models_balanced.csv", index=False)

summary = (all_results
           .dropna(subset=["accuracy"])
           .groupby(["feature_set","model"])[["accuracy","precision","recall","f1","auc","pr_auc"]]
           .mean(numeric_only=True)
           .reset_index()
           .sort_values(["feature_set","model"]))
summary.to_csv(out_dir / "occupied_sensor_ablation_summary_4models_balanced.csv", index=False)
display(summary)
print("Saved results to:", out_dir)
```

```python
# ==================== 設定 ====================
DATA_FILE = "0919Living_labeled.csv"  # 部屋内ラベル付きデータ
TARGET = "label_room"                 # 多クラス分類の目的変数
RUN_NN = True                         # Trueで LSTM/TCN も実行
WINDOW = 60                           # LSTM/TCN の系列窓長
NN_EPOCHS = 12
NN_BATCH = 128

# ==================== 共通ユーティリティ ====================
import os, numpy as np, pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, f1_score

def load_data(file):
    df = pd.read_csv(file, low_memory=False)
    if "timestamp" in df.columns:
        try:
            df["timestamp"] = pd.to_datetime(df["timestamp"])
        except Exception:
            pass
    return df

def build_preprocess(X):
    cat_cols = [c for c in X.columns if X[c].dtype == "object"]
    num_cols = [c for c in X.columns if c not in cat_cols]
    num_pipe = Pipeline([("imputer", SimpleImputer(strategy="median")),
                         ("scaler", StandardScaler())])
    cat_pipe = Pipeline([("imputer", SimpleImputer(strategy="most_frequent")),
                         ("onehot", OneHotEncoder(handle_unknown="ignore"))])
    return ColumnTransformer([("num", num_pipe, num_cols),
                              ("cat", cat_pipe, cat_cols)])

def eval_multi(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    pre, rec, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="weighted", zero_division=0)
    return {"accuracy": acc, "precision": pre, "recall": rec, "f1": f1}

# NN 用：系列化
def build_sequences(X_arr, y_arr, window=WINDOW, step=1):
    Xs, ys = [], []
    for i in range(window, len(X_arr), step):
        Xs.append(X_arr[i-window:i, :])
        ys.append(int(y_arr[i]))
    return np.array(Xs), np.array(ys)

# ==================== NNモデル構築 ====================
def get_callbacks():
    from tensorflow.keras.callbacks import EarlyStopping
    return [EarlyStopping(monitor="val_loss", patience=3, restore_best_weights=True)]

def build_lstm(input_shape, n_classes):
    from tensorflow import keras
    from tensorflow.keras import layers
    m = keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Masking(mask_value=0.0),
        layers.LSTM(64),
        layers.Dropout(0.2),
        layers.Dense(32, activation="relu"),
        layers.Dense(n_classes, activation="softmax")
    ])
    return m

def build_tcn(input_shape, n_classes):
    from tensorflow import keras
    from tensorflow.keras import layers
    inp = layers.Input(shape=input_shape)
    x = inp
    for d in [1,2,4,8]:
        x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
        x = layers.ReLU()(x)
        x = layers.Dropout(0.2)(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(32, activation="relu")(x)
    out = layers.Dense(n_classes, activation="softmax")(x)
    return keras.Model(inp, out)

# ==================== 実行フレームワーク ====================
def run_experiment(df):
    # 目的変数
    df = df.dropna(subset=[TARGET]).copy()
    le = LabelEncoder()
    y = le.fit_transform(df[TARGET])

    # 特徴量（timestamp, Occupied, Num of People は除外）
    exclude = ["timestamp","Occupied","Num of People",TARGET]
    X = df[[c for c in df.columns if c not in exclude]]

    # train/test split
    Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.2, shuffle=True, stratify=y, random_state=42)

    preprocess = build_preprocess(Xtr)
    results = []

    # ========== RandomForest ==========
    from sklearn.ensemble import RandomForestClassifier
    rf = Pipeline([("prep", preprocess),
                   ("clf", RandomForestClassifier(n_estimators=200, random_state=42))])
    rf.fit(Xtr, ytr)
    ypred = rf.predict(Xte)
    results.append({"model":"RandomForest", **eval_multi(yte, ypred)})

    # ========== HGB ==========
    from sklearn.ensemble import HistGradientBoostingClassifier
    Xtr_mat = preprocess.fit_transform(Xtr); Xte_mat = preprocess.transform(Xte)
    hgb = HistGradientBoostingClassifier(random_state=42)
    hgb.fit(Xtr_mat, ytr)
    ypred = hgb.predict(Xte_mat)
    results.append({"model":"SklearnHGB(XGB-like)", **eval_multi(yte, ypred)})

    # ========== LSTM / TCN ==========
    if RUN_NN:
        import tensorflow as tf
        from tensorflow import keras

        X_all = preprocess.fit_transform(X)
        if hasattr(X_all, "toarray"):
            X_all = X_all.toarray()

        X_seq, y_seq = build_sequences(X_all, y, window=WINDOW, step=1)
        n_classes = len(np.unique(y))

        # split train/test
        n = int(len(X_seq)*0.8)
        Xtr_seq, Xte_seq = X_seq[:n], X_seq[n:]
        ytr_seq, yte_seq = y_seq[:n], y_seq[n:]

        # one-hot
        ytr_oh = keras.utils.to_categorical(ytr_seq, num_classes=n_classes)
        yte_oh = keras.utils.to_categorical(yte_seq, num_classes=n_classes)

        input_shape = (Xtr_seq.shape[1], Xtr_seq.shape[2])

        # LSTM
        lstm = build_lstm(input_shape, n_classes)
        lstm.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
        lstm.fit(Xtr_seq, ytr_oh, validation_split=0.2, epochs=NN_EPOCHS,
                 batch_size=NN_BATCH, callbacks=get_callbacks(), verbose=0)
        ypred = lstm.predict(Xte_seq).argmax(axis=1)
        results.append({"model":"LSTM", **eval_multi(yte_seq, ypred)})

        # TCN
        tcn = build_tcn(input_shape, n_classes)
        tcn.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])
        tcn.fit(Xtr_seq, ytr_oh, validation_split=0.2, epochs=NN_EPOCHS,
                batch_size=NN_BATCH, callbacks=get_callbacks(), verbose=0)
        ypred = tcn.predict(Xte_seq).argmax(axis=1)
        results.append({"model":"TCN", **eval_multi(yte_seq, ypred)})

    return pd.DataFrame(results)

# ==================== 実行 ====================
df = load_data(DATA_FILE)
results = run_experiment(df)
print(results)
```

```python
# ==================== Config ====================
DATA_FILE = "0919Living_labeled.csv"   # アップロードしたCSV
TARGET = "label_room"                  # 位置ラベル（多クラス）
SEED = 42

# NN
RUN_NN = True
WINDOW = 90            # 窓長（環境+動きの遅速に合わせて 60〜180 を試す）
STEP = 3               # シーケンス作成時のステップ（データ増強）
NN_EPOCHS = 30
NN_BATCH = 128
LABEL_SMOOTHING = 0.05

# 依存
import os, random, numpy as np, pandas as pd
from pathlib import Path
random.seed(SEED); np.random.seed(SEED)

# ==================== Load ====================
df = pd.read_csv(DATA_FILE, low_memory=False)
if "timestamp" in df.columns:
    try:
        df["timestamp"] = pd.to_datetime(df["timestamp"])
        df = df.sort_values("timestamp").reset_index(drop=True)
    except Exception:
        pass

# 目的変数の存在チェック
if TARGET not in df.columns:
    raise ValueError(f"'{TARGET}' 列が見つかりません。columns={list(df.columns)[:20]} ...")

# ラベル欠損を除外
df = df.dropna(subset=[TARGET]).copy()

# 学習に使わない列を除外
exclude = {"timestamp","Occupied","Num of People", TARGET}
feat_cols = [c for c in df.columns if c not in exclude]
Xall = df[feat_cols].copy()
y_raw = df[TARGET].astype(str).copy()

# ==================== 前処理（ColumnTransformer） ====================
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer

cat_cols = [c for c in Xall.columns if Xall[c].dtype=="object"]
num_cols = [c for c in Xall.columns if c not in cat_cols]

num_pipe = Pipeline([("imputer", SimpleImputer(strategy="median")),
                     ("scaler", StandardScaler())])
cat_pipe = Pipeline([("imputer", SimpleImputer(strategy="most_frequent")),
                     ("onehot", OneHotEncoder(handle_unknown="ignore"))])

pre = ColumnTransformer([("num", num_pipe, num_cols),
                         ("cat", cat_pipe, cat_cols)])

# ==================== 時系列分割（60/20/20, 時間順） ====================
n = len(df)
n_tr = int(n*0.6); n_dv = int(n*0.8)

X_tr_df, X_dv_df, X_te_df = Xall.iloc[:n_tr], Xall.iloc[n_tr:n_dv], Xall.iloc[n_dv:]
y_tr_raw, y_dv_raw, y_te_raw = y_raw.iloc[:n_tr], y_raw.iloc[n_tr:n_dv], y_raw.iloc[n_dv:]

# ラベルエンコード（全体の語彙でfit）
le = LabelEncoder()
le.fit(y_raw)
y_tr = le.transform(y_tr_raw)
y_dv = le.transform(y_dv_raw)
y_te = le.transform(y_te_raw)
classes = le.classes_
n_classes = len(classes)
print("Classes:", list(classes))

# ==================== 2系統：ツリー系（RF/HGB）とNN系 ====================
# -------- ツリー系：RF / HGB --------
from sklearn.metrics import classification_report, accuracy_score, f1_score, precision_recall_fscore_support, confusion_matrix
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier

def eval_multi(y_true, y_pred, labels=None):
    acc = accuracy_score(y_true, y_pred)
    pre_w, rec_w, f1_w, _ = precision_recall_fscore_support(y_true, y_pred, average="weighted", zero_division=0)
    pre_m, rec_m, f1_m, _ = precision_recall_fscore_support(y_true, y_pred, average="macro", zero_division=0)
    return {
        "accuracy": acc,
        "precision_weighted": pre_w, "recall_weighted": rec_w, "f1_weighted": f1_w,
        "precision_macro": pre_m, "recall_macro": rec_m, "f1_macro": f1_m
    }

results = []

# RF
rf = Pipeline([("pre", pre),
               ("clf", RandomForestClassifier(n_estimators=400, max_depth=None,
                                             random_state=SEED, n_jobs=-1))])
rf.fit(X_tr_df, y_tr)
y_hat = rf.predict(X_te_df)
results.append({"model":"RandomForest", **eval_multi(y_te, y_hat)})

# HGB (XGB-like)
Xtr_mat = pre.fit_transform(X_tr_df)
Xdv_mat = pre.transform(X_dv_df)
Xte_mat = pre.transform(X_te_df)
hgb = HistGradientBoostingClassifier(random_state=SEED)
hgb.fit(Xtr_mat, y_tr)
y_hat = hgb.predict(Xte_mat)
results.append({"model":"SklearnHGB(XGB-like)", **eval_multi(y_te, y_hat)})

# -------- NN系：LSTM / TCN --------
def build_sequences(X_mat, y_arr, window=WINDOW, step=STEP):
    Xs, ys = [], []
    for i in range(window, len(X_mat), step):
        Xs.append(X_mat[i-window:i, :])
        ys.append(int(y_arr[i]))
    return np.array(Xs), np.array(ys)

def balance_sequences(Xs, ys, max_per_class=None):
    # 各クラスの最小数に合わせてダウンサンプル
    import numpy as np
    Xs, ys = np.asarray(Xs), np.asarray(ys)
    uniq, counts = np.unique(ys, return_counts=True)
    k = counts.min()
    if max_per_class is not None:
        k = min(k, max_per_class)
    idxs = []
    for c in uniq:
        cand = np.where(ys==c)[0]
        if len(cand) >= k:
            sel = np.random.choice(cand, k, replace=False)
            idxs.append(sel)
    if len(idxs)==0:
        return Xs, ys
    sel = np.concatenate(idxs)
    np.random.shuffle(sel)
    return Xs[sel], ys[sel]

if RUN_NN:
    import tensorflow as tf
    from tensorflow import keras
    tf.random.set_seed(SEED)

    # 全データを同じ前処理器で変換（fitはtrainのみに合わせるのが本則だが、ここは train fit→全体transform）
    pre_nn = ColumnTransformer([("num", num_pipe, num_cols),
                                ("cat", cat_pipe, cat_cols)])
    Xtr_mat_nn = pre_nn.fit_transform(X_tr_df)
    Xdv_mat_nn = pre_nn.transform(X_dv_df)
    Xte_mat_nn = pre_nn.transform(X_te_df)

    if hasattr(Xtr_mat_nn, "toarray"):
        Xtr_mat_nn = Xtr_mat_nn.toarray()
        Xdv_mat_nn = Xdv_mat_nn.toarray()
        Xte_mat_nn = Xte_mat_nn.toarray()

    # シーケンス化（train/dev/test）
    Xtr_seq, ytr_seq = build_sequences(Xtr_mat_nn, y_tr, window=WINDOW, step=STEP)
    Xdv_seq, ydv_seq = build_sequences(np.vstack([Xtr_mat_nn[-WINDOW+1:,:], Xdv_mat_nn]), np.concatenate([y_tr[-WINDOW+1:], y_dv]), window=WINDOW, step=STEP)
    Xte_seq, yte_seq = build_sequences(np.vstack([Xdv_mat_nn[-WINDOW+1:,:], Xte_mat_nn]), np.concatenate([y_dv[-WINDOW+1:], y_te]), window=WINDOW, step=STEP)

    # 学習バランス化
    Xtr_bal, ytr_bal = balance_sequences(Xtr_seq, ytr_seq, max_per_class=8000)

    # one-hot & class weight
    ytr_oh = keras.utils.to_categorical(ytr_bal, num_classes=n_classes)
    ydv_oh = keras.utils.to_categorical(ydv_seq, num_classes=n_classes)

    # class weights（頻度の逆数）
    # （この行より前はそのまま）ytr_bal / ydv_oh など作成済みを想定

    # ---- class weights を ytr_bal の実在クラスで安全に作る ----
    from sklearn.utils.class_weight import compute_class_weight
    uniq_in_train = np.unique(ytr_bal)

    # もし系列化後にクラスが1種以下なら、NNはスキップ
    if len(uniq_in_train) < 2:
        print("[NN skipped] Not enough classes in balanced train sequences:", uniq_in_train)
        # ここで return しない。RF/HGB の結果は既に results に入っているはず
    else:
        cw_present = compute_class_weight(class_weight="balanced",
                                          classes=uniq_in_train, y=ytr_bal)
        # 欠けているクラスは 1.0 で埋める
        class_weight = {int(c): 1.0 for c in range(n_classes)}
        for c, w in zip(uniq_in_train, cw_present):
            class_weight[int(c)] = float(w)

        input_shape = (Xtr_bal.shape[1], Xtr_bal.shape[2])

        # --- LSTM ---
        lstm = build_lstm(input_shape, n_classes)
        lstm.compile(optimizer=keras.optimizers.Adam(1e-3),
                    loss=keras.losses.CategoricalCrossentropy(label_smoothing=LABEL_SMOOTHING),
                    metrics=["accuracy"])
        cbs = [
            keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=2, min_lr=1e-5, verbose=1),
            keras.callbacks.EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True, verbose=1),
        ]
        lstm.fit(Xtr_bal, ytr_oh, validation_data=(Xdv_seq, ydv_oh),
                epochs=NN_EPOCHS, batch_size=NN_BATCH,
                class_weight=class_weight, verbose=1, callbacks=cbs)
        yhat = lstm.predict(Xte_seq, verbose=0).argmax(axis=1)
        results.append({"model":"LSTM", **eval_multi(yte_seq, yhat)})

        # --- TCN ---
        tcn = build_tcn(input_shape, n_classes)
        tcn.compile(optimizer=keras.optimizers.Adam(1e-3),
                    loss=keras.losses.CategoricalCrossentropy(label_smoothing=LABEL_SMOOTHING),
                    metrics=["accuracy"])
        tcn.fit(Xtr_bal, ytr_oh, validation_data=(Xdv_seq, ydv_oh),
                epochs=NN_EPOCHS, batch_size=NN_BATCH,
                class_weight=class_weight, verbose=1, callbacks=cbs)
        yhat = tcn.predict(Xte_seq, verbose=0).argmax(axis=1)
        results.append({"model":"TCN", **eval_multi(yte_seq, yhat)})


# ==================== 出力（表と混同行列） ====================
res_df = pd.DataFrame(results)
display(res_df)

# 混同行列（ベストモデルを例に）
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
best_idx = res_df["f1_weighted"].astype(float).idxmax()
best_name = res_df.loc[best_idx, "model"]
print("Best (by weighted F1):", best_name)

# 予測を取り直す（可視化用）
if best_name == "RandomForest":
    y_pred = rf.predict(X_te_df)
elif best_name == "SklearnHGB(XGB-like)":
    y_pred = hgb.predict(Xte_mat)
elif best_name == "LSTM":
    y_pred = lstm.predict(Xte_seq, verbose=0).argmax(axis=1)
    # yte_seq / classes はエンコード済
    # 以降は yte_seq, y_pred をそのまま利用
    fig, ax = plt.subplots(figsize=(6,5))
    ConfusionMatrixDisplay(confusion_matrix(yte_seq, y_pred), display_labels=classes).plot(ax=ax, cmap="Blues", colorbar=False)
    plt.title(f"Confusion Matrix - {best_name}")
    plt.tight_layout(); plt.show()
else:
    y_pred = tcn.predict(Xte_seq, verbose=0).argmax(axis=1)
    fig, ax = plt.subplots(figsize=(6,5))
    ConfusionMatrixDisplay(confusion_matrix(yte_seq, y_pred), display_labels=classes).plot(ax=ax, cmap="Blues", colorbar=False)
    plt.title(f"Confusion Matrix - {best_name}")
    plt.tight_layout(); plt.show()

# ツリー系の混同行列も保存（参考）
if best_name in ["RandomForest","SklearnHGB(XGB-like)"]:
    fig, ax = plt.subplots(figsize=(6,5))
    ConfusionMatrixDisplay(confusion_matrix(y_te, y_pred), display_labels=classes).plot(ax=ax, cmap="Blues", colorbar=False)
    plt.title(f"Confusion Matrix - {best_name}")
    plt.tight_layout(); plt.show()
```

```python
# ==============================
# Room Position Classification (Multiclass)
#  - target: label_room (kitchen / living_east / living_west / out of room)
#  - models: RandomForest / HistGB / LSTM / TCN
#  - robust to non-numeric cols, class imbalance, time leakage
# ==============================

# ====== 設定 ======
DATA_FILE = "0919Living_labeled.csv"
TARGET = "label_room"
TIMESTAMP_COL = "timestamp"   # あれば時系列ソートに使う（特徴には入れない）

SEED = 42
WINDOW = 90   # シーケンス窓長（60〜120程度で要調整）
STEP   = 3    # シーケンス生成のステップ（1で密、3で間引き）
EPOCHS = 30
BATCH  = 128

import os, random, numpy as np, pandas as pd
random.seed(SEED); np.random.seed(SEED)

# ====== データ読み込み ======
df = pd.read_csv(DATA_FILE, low_memory=False)
if TIMESTAMP_COL in df.columns:
    try:
        df[TIMESTAMP_COL] = pd.to_datetime(df[TIMESTAMP_COL])
        df = df.sort_values(TIMESTAMP_COL).reset_index(drop=True)
    except Exception:
        pass

# 目的変数の欠損除去
df = df.dropna(subset=[TARGET]).copy()

# ====== 特徴量とラベル ======
# リーク可能列は除外（必要に応じて追加してOK）
exclude = {TARGET, TIMESTAMP_COL, "Occupied", "Num of People"}
feat_cols = [c for c in df.columns if c not in exclude]
X_df = df[feat_cols].copy()
y_raw = df[TARGET].astype(str).copy()

# ラベルエンコード
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_all = le.fit_transform(y_raw)
classes = le.classes_.tolist()
n_classes = len(classes)
print("Classes:", classes)

# ====== 前処理（数値: median+scale / 文字列: most_frequent+OneHot） ======
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

cat_cols = [c for c in X_df.columns if X_df[c].dtype == "object"]
num_cols = [c for c in X_df.columns if c not in cat_cols]

num_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
])
cat_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore")),
])
pre = ColumnTransformer([
    ("num", num_pipe, num_cols),
    ("cat", cat_pipe, cat_cols),
])

# ====== 層化分割（Tabular: train/dev/test = 60/20/20, ランダム層化） ======
from sklearn.model_selection import train_test_split
X_tr_df, X_tmp_df, y_tr, y_tmp = train_test_split(
    X_df, y_all, test_size=0.4, random_state=SEED, stratify=y_all, shuffle=True
)
X_dv_df, X_te_df, y_dv, y_te = train_test_split(
    X_tmp_df, y_tmp, test_size=0.5, random_state=SEED, stratify=y_tmp, shuffle=True
)

# Tabular行列化
Xtr_tab = pre.fit_transform(X_tr_df)
Xdv_tab = pre.transform(X_dv_df)
Xte_tab = pre.transform(X_te_df)

if hasattr(Xtr_tab, "toarray"):
    Xtr_tab = Xtr_tab.toarray(); Xdv_tab = Xdv_tab.toarray(); Xte_tab = Xte_tab.toarray()

# ====== 指標関数 ======
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
def eval_multi(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    pw, rw, fw, _ = precision_recall_fscore_support(y_true, y_pred, average="weighted", zero_division=0)
    pm, rm, fm, _ = precision_recall_fscore_support(y_true, y_pred, average="macro", zero_division=0)
    return {
        "accuracy": acc,
        "precision_weighted": pw, "recall_weighted": rw, "f1_weighted": fw,
        "precision_macro": pm, "recall_macro": rm, "f1_macro": fm,
    }

# ====== 1) RandomForest ======
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(
    n_estimators=400, max_depth=None, min_samples_leaf=1,
    class_weight="balanced", random_state=SEED, n_jobs=-1
)
rf.fit(Xtr_tab, y_tr)
y_rf = rf.predict(Xte_tab)

# ====== 2) HistGradientBoosting (XGB-like) ======
from sklearn.ensemble import HistGradientBoostingClassifier
hgb = HistGradientBoostingClassifier(random_state=SEED)
hgb.fit(Xtr_tab, y_tr)
y_hgb = hgb.predict(Xte_tab)

# ====== NN用：シーケンス作成 ======
# 連結してから窓作成。ただし学習/検証/テストの境界を跨いだ窓は切り分けで吸収
def stack_for_sequence(Xtr, Xdv, Xte, ytr, ydv, yte):
    X_all = np.vstack([Xtr, Xdv, Xte])
    y_all = np.concatenate([ytr, ydv, yte])
    i_tr_end = len(Xtr)
    i_dv_end = len(Xtr) + len(Xdv)
    return X_all, y_all, i_tr_end, i_dv_end

X_all_mat, y_all_seq, i_tr_end, i_dv_end = stack_for_sequence(Xtr_tab, Xdv_tab, Xte_tab, y_tr, y_dv, y_te)

def make_sequences(X, y, window=WINDOW, step=STEP):
    Xs, ys = [], []
    for i in range(window, len(X), step):
        Xs.append(X[i-window:i, :])
        ys.append(int(y[i]))
    return np.array(Xs), np.array(ys)

X_seq, y_seq = make_sequences(X_all_mat, y_all_seq, WINDOW, STEP)

# シーケンスを train/dev/test に切り分け（窓終端位置で判定）
ends = np.arange(WINDOW, len(X_all_mat), STEP)
mask_tr = ends <= i_tr_end
mask_dv = (ends > i_tr_end) & (ends <= i_dv_end)
mask_te = ends > i_dv_end

Xtr_seq, ytr_seq = X_seq[mask_tr], y_seq[mask_tr]
Xdv_seq, ydv_seq = X_seq[mask_dv], y_seq[mask_dv]
Xte_seq, yte_seq = X_seq[mask_te], y_seq[mask_te]
print("Seq shapes:", Xtr_seq.shape, Xdv_seq.shape, Xte_seq.shape)

# クラス重み（実際に学習に含まれるクラスのみで安全に算出）
from sklearn.utils.class_weight import compute_class_weight
present = np.unique(ytr_seq)
if len(present) >= 2:
    cw_present = compute_class_weight(class_weight="balanced", classes=present, y=ytr_seq)
    class_weight = {int(c): float(w) for c, w in zip(present, cw_present)}
else:
    class_weight = None
    print("[NN Notice] Only one class present in training sequences; skip class_weight.")

# ====== 3) LSTM ======
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
tf.random.set_seed(SEED)

input_shape = (Xtr_seq.shape[1], Xtr_seq.shape[2])

def build_lstm(input_shape, n_classes):
    m = keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Masking(mask_value=0.0),
        layers.LSTM(64, return_sequences=True, dropout=0.2),
        layers.LSTM(64, dropout=0.2),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(n_classes, activation="softmax")
    ])
    # ★ label_smoothing を使わない互換構成
    m.compile(optimizer=keras.optimizers.Adam(1e-3),
              loss=keras.losses.SparseCategoricalCrossentropy(),
              metrics=["accuracy"])
    return m

def fit_nn(model, Xtr, ytr, Xdv, ydv, cw):
    cbs = [
        keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=2, min_lr=1e-5, verbose=1),
        keras.callbacks.EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True, verbose=1),
    ]
    return model.fit(Xtr, ytr, validation_data=(Xdv, ydv),
                     epochs=EPOCHS, batch_size=BATCH, verbose=1,
                     class_weight=cw, callbacks=cbs)

lstm = build_lstm(input_shape, n_classes)
hist_lstm = fit_nn(lstm, Xtr_seq, ytr_seq, Xdv_seq, ydv_seq, class_weight)
y_lstm = lstm.predict(Xte_seq, verbose=0).argmax(axis=1)

# ====== 4) TCN ======
def build_tcn(input_shape, n_classes):
    inp = layers.Input(shape=input_shape)
    x = inp
    for d in [1, 2, 4, 8, 16]:
        res = x
        x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
        x = layers.BatchNormalization()(x)
        x = layers.ReLU()(x)
        x = layers.Dropout(0.2)(x)
        x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
        x = layers.BatchNormalization()(x)
        if res.shape[-1] != x.shape[-1]:
            res = layers.Conv1D(64, 1, padding="same")(res)
        x = layers.Add()([x, res])
        x = layers.ReLU()(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(64, activation="relu")(x)
    x = layers.Dropout(0.3)(x)
    out = layers.Dense(n_classes, activation="softmax")(x)
    m = keras.Model(inp, out)
    # ★ label_smoothing を使わない互換構成
    m.compile(optimizer=keras.optimizers.Adam(1e-3),
              loss=keras.losses.SparseCategoricalCrossentropy(),
              metrics=["accuracy"])
    return m

tcn = build_tcn(input_shape, n_classes)
hist_tcn = fit_nn(tcn, Xtr_seq, ytr_seq, Xdv_seq, ydv_seq, class_weight)
y_tcn = tcn.predict(Xte_seq, verbose=0).argmax(axis=1)

# ====== 結果集計 ======
results = []
results.append({"model":"RandomForest", **eval_multi(y_te,    y_rf)})
results.append({"model":"SklearnHGB(XGB-like)", **eval_multi(y_te,    y_hgb)})
results.append({"model":"LSTM", **eval_multi(yte_seq, y_lstm)})
results.append({"model":"TCN",  **eval_multi(yte_seq, y_tcn)})

res_df = pd.DataFrame(results)
display(res_df)

# ====== 混同行列（安全表示：出現クラス数に自動追従） ======
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, classification_report

def plot_cm(y_true, y_pred, class_names, title, savepath=None):
    labs = np.unique(np.concatenate([np.unique(y_true), np.unique(y_pred)]))
    cm = confusion_matrix(y_true, y_pred, labels=labs)
    fig, ax = plt.subplots(figsize=(6,5))
    ConfusionMatrixDisplay(cm, display_labels=[class_names[i] for i in labs]).plot(ax=ax, cmap="Blues", colorbar=False)
    ax.set_title(title)
    plt.tight_layout()
    if savepath: plt.savefig(savepath, dpi=150)
    plt.show()
    print(classification_report(y_true, y_pred, labels=labs,
                                target_names=[class_names[i] for i in labs], zero_division=0))

plot_cm(y_te,     y_rf,   classes, "CM - RandomForest")
plot_cm(y_te,     y_hgb,  classes, "CM - HGB")
plot_cm(yte_seq,  y_lstm, classes, "CM - LSTM")
plot_cm(yte_seq,  y_tcn,  classes, "CM - TCN")
```

```python
# ============================================
# Room Position Classification (Multiclass) – Integrated 4-model pipeline
# target: label_room  (e.g., kitchen / living_east / living_west / out of room)
# models: RandomForest / HistGB / LSTM / TCN
# design: time-ordered split (60/20/20) unified across models, time features, class balance for sequences,
#         robust preprocessing (numeric: median+scale, categorical: most_frequent+onehot),
#         safe confusion matrices and full metrics.
# ============================================

# ===== Config =====
DATA_FILE   = "0919Living_labeled.csv"
TARGET      = "label_room"
TIMESTAMP   = "timestamp"    # if present: used for sorting only (NOT as feature)
EXCLUDE     = {TARGET, TIMESTAMP, "Occupied", "Num of People"}  # add leaking columns here

SEED        = 42
WINDOW      = 60   # sequence window length (try 60/90/120)
STEP        = 1    # step when slicing sequences (1 = dense)
EPOCHS      = 30
BATCH       = 128

import os, random, numpy as np, pandas as pd
random.seed(SEED); np.random.seed(SEED)

# ===== Load =====
df = pd.read_csv(DATA_FILE, low_memory=False)
if TIMESTAMP in df.columns:
    try:
        df[TIMESTAMP] = pd.to_datetime(df[TIMESTAMP])
        df = df.sort_values(TIMESTAMP).reset_index(drop=True)
    except Exception:
        pass

# drop rows with missing label
df = df.dropna(subset=[TARGET]).copy()

# ===== Features & Target =====
feat_cols = [c for c in df.columns if c not in EXCLUDE]
X_df = df[feat_cols].copy()
y_raw = df[TARGET].astype(str).copy()

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_all = le.fit_transform(y_raw)
CLASSES = le.classes_.tolist()
N_CLASSES = len(CLASSES)
print("Classes:", CLASSES)

# ===== Preprocess: numeric (median+scale) / categorical (most_frequent+onehot) =====
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

cat_cols = [c for c in X_df.columns if X_df[c].dtype == "object"]
num_cols = [c for c in X_df.columns if c not in cat_cols]

num_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler",  StandardScaler()),
])
cat_pipe = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot",  OneHotEncoder(handle_unknown="ignore")),
])
pre = ColumnTransformer([
    ("num", num_pipe, num_cols),
    ("cat", cat_pipe, cat_cols),
])

# ===== Time-ordered split (60/20/20) – unified for ALL models =====
N = len(df)
N_TR = int(N * 0.60)
N_DV = int(N * 0.80)

X_tr_df, X_dv_df, X_te_df = X_df.iloc[:N_TR], X_df.iloc[N_TR:N_DV], X_df.iloc[N_DV:]
y_tr,     y_dv,     y_te   = y_all[:N_TR],   y_all[N_TR:N_DV],   y_all[N_DV:]

# Fit on train, transform dev/test
Xtr_tab = pre.fit_transform(X_tr_df)
Xdv_tab = pre.transform(X_dv_df)
Xte_tab = pre.transform(X_te_df)

# materialize to dense (for NN & time features)
if hasattr(Xtr_tab, "toarray"):
    Xtr_tab = Xtr_tab.toarray()
    Xdv_tab = Xdv_tab.toarray()
    Xte_tab = Xte_tab.toarray()

# ===== Add time features (1st diff, 2nd diff, moving average) =====
def add_time_feats(X, win=9):
    # X: (T, F). returns concat [X, d1, d2, MA]
    T, F = X.shape
    d1 = np.vstack([np.zeros((1, F)), np.diff(X, n=1, axis=0)])
    d2 = np.vstack([np.zeros((2, F)), np.diff(X, n=2, axis=0)])
    ma = np.zeros_like(X)
    k = win // 2
    for t in range(T):
        s = max(0, t - k); e = min(T, t + k + 1)
        ma[t] = X[s:e].mean(axis=0)
    return np.hstack([X, d1, d2, ma])

Xtr_tf = add_time_feats(Xtr_tab, win=9)
Xdv_tf = add_time_feats(Xdv_tab, win=9)
Xte_tf = add_time_feats(Xte_tab, win=9)

# ===== Classical models (on tabular with time-ordered split) =====
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
def eval_multi(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    pw, rw, fw, _ = precision_recall_fscore_support(y_true, y_pred, average="weighted", zero_division=0)
    pm, rm, fm, _ = precision_recall_fscore_support(y_true, y_pred, average="macro", zero_division=0)
    return {
        "accuracy": acc,
        "precision_weighted": pw, "recall_weighted": rw, "f1_weighted": fw,
        "precision_macro": pm, "recall_macro": rm, "f1_macro": fm,
    }

from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier

rf = RandomForestClassifier(
    n_estimators=500, max_depth=None, min_samples_leaf=1,
    class_weight="balanced", random_state=SEED, n_jobs=-1
)
rf.fit(Xtr_tf, y_tr)
y_rf = rf.predict(Xte_tf)

hgb = HistGradientBoostingClassifier(random_state=SEED)
hgb.fit(Xtr_tf, y_tr)
y_hgb = hgb.predict(Xte_tf)

# ===== Build sequences from time-featured matrices =====
def make_sequences(X, y, window=WINDOW, step=STEP):
    Xs, ys = [], []
    for i in range(window, len(X), step):
        Xs.append(X[i-window:i, :])
        ys.append(int(y[i]))
    return np.array(Xs), np.array(ys)

# concat to preserve global chronology then cut by end indices
X_all_mat = np.vstack([Xtr_tf, Xdv_tf, Xte_tf])
y_all_seq = np.concatenate([y_tr, y_dv, y_te])
i_tr_end, i_dv_end = len(Xtr_tf), len(Xtr_tf) + len(Xdv_tf)

X_seq, y_seq = make_sequences(X_all_mat, y_all_seq, WINDOW, STEP)
ends = np.arange(WINDOW, len(X_all_mat), STEP)
mask_tr = ends <= i_tr_end
mask_dv = (ends > i_tr_end) & (ends <= i_dv_end)
mask_te = ends > i_dv_end

Xtr_seq, ytr_seq = X_seq[mask_tr], y_seq[mask_tr]
Xdv_seq, ydv_seq = X_seq[mask_dv], y_seq[mask_dv]
Xte_seq, yte_seq = X_seq[mask_te], y_seq[mask_te]
print("Seq shapes:", Xtr_seq.shape, Xdv_seq.shape, Xte_seq.shape)

# balance training sequences across classes
def balance_sequences(Xs, ys, max_per_class=None, seed=SEED):
    rng = np.random.default_rng(seed)
    Xs, ys = np.asarray(Xs), np.asarray(ys)
    uniq, cnt = np.unique(ys, return_counts=True)
    k = cnt.min()
    if max_per_class is not None:
        k = min(k, max_per_class)
    idxs = []
    for c in uniq:
        cand = np.where(ys == c)[0]
        sel = rng.choice(cand, k, replace=False)
        idxs.append(sel)
    sel = np.concatenate(idxs)
    rng.shuffle(sel)
    return Xs[sel], ys[sel]

Xtr_seq_bal, ytr_seq_bal = balance_sequences(Xtr_seq, ytr_seq, max_per_class=8000)

# class weights for present classes only
from sklearn.utils.class_weight import compute_class_weight
present = np.unique(ytr_seq_bal)
if len(present) >= 2:
    cw_present = compute_class_weight(class_weight="balanced", classes=present, y=ytr_seq_bal)
    CLASS_WEIGHT = {int(c): float(w) for c, w in zip(present, cw_present)}
else:
    CLASS_WEIGHT = None
    print("[NN Notice] Only one class in training sequences; skip class_weight.")

# ===== LSTM / TCN (TF/Keras) =====
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
tf.random.set_seed(SEED)

INPUT_SHAPE = (Xtr_seq_bal.shape[1], Xtr_seq_bal.shape[2])

def build_lstm(input_shape, n_classes):
    m = keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Masking(mask_value=0.0),
        layers.LSTM(64, return_sequences=True, dropout=0.2),
        layers.LSTM(64, dropout=0.2),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(n_classes, activation="softmax")
    ])
    m.compile(optimizer=keras.optimizers.Adam(1e-3),
              loss=keras.losses.SparseCategoricalCrossentropy(),
              metrics=["accuracy"])
    return m

def build_tcn(input_shape, n_classes):
    inp = layers.Input(shape=input_shape)
    x = inp
    for d in [1,2,4,8,16]:
        res = x
        x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
        x = layers.BatchNormalization()(x)
        x = layers.ReLU()(x)
        x = layers.Dropout(0.2)(x)
        x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
        x = layers.BatchNormalization()(x)
        if res.shape[-1] != x.shape[-1]:
            res = layers.Conv1D(64, 1, padding="same")(res)
        x = layers.Add()([x, res]); x = layers.ReLU()(x)
    x = layers.LayerNormalization()(x)           # stabilization
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(64, activation="relu")(x)
    x = layers.Dropout(0.3)(x)
    out = layers.Dense(n_classes, activation="softmax")(x)
    m = keras.Model(inp, out)
    m.compile(optimizer=keras.optimizers.Adam(1e-3),
              loss=keras.losses.SparseCategoricalCrossentropy(),
              metrics=["accuracy"])
    return m

def fit_nn(model, Xtr, ytr, Xdv, ydv, cw):
    cbs = [
        keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=2, min_lr=1e-5, verbose=1),
        keras.callbacks.EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True, verbose=1),
    ]
    return model.fit(Xtr, ytr, validation_data=(Xdv, ydv),
                     epochs=EPOCHS, batch_size=BATCH, verbose=1,
                     class_weight=cw, callbacks=cbs)

# train & predict
lstm = build_lstm(INPUT_SHAPE, N_CLASSES)
_ = fit_nn(lstm, Xtr_seq_bal, ytr_seq_bal, Xdv_seq, ydv_seq, CLASS_WEIGHT)
y_lstm = lstm.predict(Xte_seq, verbose=0).argmax(axis=1)

tcn = build_tcn(INPUT_SHAPE, N_CLASSES)
_ = fit_nn(tcn, Xtr_seq_bal, ytr_seq_bal, Xdv_seq, ydv_seq, CLASS_WEIGHT)
y_tcn = tcn.predict(Xte_seq, verbose=0).argmax(axis=1)

# ===== Results table =====
results = []
from pprint import pprint
results.append({"model":"RandomForest",            **eval_multi(y_te,    y_rf )})
results.append({"model":"SklearnHGB(XGB-like)",    **eval_multi(y_te,    y_hgb)})
results.append({"model":"LSTM",                    **eval_multi(yte_seq, y_lstm)})
results.append({"model":"TCN",                     **eval_multi(yte_seq, y_tcn)})

res_df = pd.DataFrame(results)
display(res_df)

# ===== Safe confusion matrices =====
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix, classification_report

def plot_cm(y_true, y_pred, class_names, title, savepath=None):
    labs = np.unique(np.concatenate([np.unique(y_true), np.unique(y_pred)]))
    cm = confusion_matrix(y_true, y_pred, labels=labs)
    fig, ax = plt.subplots(figsize=(6,5))
    ConfusionMatrixDisplay(cm, display_labels=[class_names[i] for i in labs]).plot(ax=ax, cmap="Blues", colorbar=False)
    ax.set_title(title)
    plt.tight_layout()
    if savepath: plt.savefig(savepath, dpi=150)
    plt.show()
    print(classification_report(y_true, y_pred, labels=labs,
                                target_names=[class_names[i] for i in labs],
                                zero_division=0))

plot_cm(y_te,    y_rf,   CLASSES, "CM - RandomForest")
plot_cm(y_te,    y_hgb,  CLASSES, "CM - HGB")
plot_cm(yte_seq, y_lstm, CLASSES, "CM - LSTM")
plot_cm(yte_seq, y_tcn,  CLASSES, "CM - TCN")
```

```python
# ============================================
# Room Position Classification – Fair Sequence-Stratified Split Version
#   ・シーケンス（窓）を先に作り、その終端サンプルで層化分割
#   ・全モデル（RF/HGB/LSTM/TCN）を同じサンプル集合で比較
# ============================================

# ===== Config =====
DATA_FILE   = "0919Living_labeled.csv"
TARGET      = "label_room"
TIMESTAMP   = "timestamp"    # あれば並び替えにだけ使う
EXCLUDE     = {TARGET, TIMESTAMP, "Occupied", "Num of People"}

SEED        = 42
WINDOW      = 60   # 窓長
STEP        = 1    # 窓ステップ
EPOCHS      = 30
BATCH       = 128

import os, random, numpy as np, pandas as pd
random.seed(SEED); np.random.seed(SEED)

# ===== Load =====
df = pd.read_csv(DATA_FILE, low_memory=False)
if TIMESTAMP in df.columns:
    try:
        df[TIMESTAMP] = pd.to_datetime(df[TIMESTAMP])
        df = df.sort_values(TIMESTAMP).reset_index(drop=True)
    except Exception:
        pass
df = df.dropna(subset=[TARGET]).copy()

# ===== Features / Label =====
feat_cols = [c for c in df.columns if c not in EXCLUDE]
X_df = df[feat_cols].copy()
y_raw = df[TARGET].astype(str).copy()

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_all = le.fit_transform(y_raw)
CLASSES = le.classes_.tolist()
N_CLASSES = len(CLASSES)
print("Classes:", CLASSES)

# ===== Preprocess (numeric/categorical) =====
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

cat_cols = [c for c in X_df.columns if X_df[c].dtype == "object"]
num_cols = [c for c in X_df.columns if c not in cat_cols]

num_pipe = Pipeline([("imputer", SimpleImputer(strategy="median")),
                     ("scaler",  StandardScaler())])
cat_pipe = Pipeline([("imputer", SimpleImputer(strategy="most_frequent")),
                     ("onehot",  OneHotEncoder(handle_unknown="ignore"))])

pre = ColumnTransformer([("num", num_pipe, num_cols),
                         ("cat", cat_pipe, cat_cols)])

# ★ 前処理は「全体に fit」しない（リーク防止）
#   → 時系列の先頭 60% を train とみなし fit、残りは transform のみ。
N = len(df)
N_TR_FIT = int(N * 0.60)
Xtr_fit = pre.fit_transform(X_df.iloc[:N_TR_FIT])
Xte_part = pre.transform(X_df.iloc[N_TR_FIT:])

# ただし、この後は「シーケンス単位で層化分割」を行うため、
# 前処理済みの “全体行列” を作る（train fit 済みの pre を使って transform のみ）
X_all_mat = np.vstack([Xtr_fit, pre.transform(X_df.iloc[N_TR_FIT:])])
if hasattr(X_all_mat, "toarray"):
    X_all_mat = X_all_mat.toarray()

# ===== 時間特徴（差分・移動平均など） =====
def add_time_feats(X, win=9):
    T, F = X.shape
    d1 = np.vstack([np.zeros((1, F)), np.diff(X, n=1, axis=0)])
    d2 = np.vstack([np.zeros((2, F)), np.diff(X, n=2, axis=0)])
    ma = np.zeros_like(X); k = win // 2
    for t in range(T):
        s = max(0, t-k); e = min(T, t+k+1)
        ma[t] = X[s:e].mean(axis=0)
    return np.hstack([X, d1, d2, ma])

X_all_tf = add_time_feats(X_all_mat, win=9)

# ===== シーケンス作成（全体→終端インデックスを得る） =====
def make_sequences(X, y, window=WINDOW, step=STEP):
    Xs, ys, ends = [], [], []
    for i in range(window, len(X), step):
        Xs.append(X[i-window:i, :])
        ys.append(int(y[i]))
        ends.append(i)  # 終端の行インデックス
    return np.array(Xs), np.array(ys), np.array(ends)

X_seq, y_seq, end_idx = make_sequences(X_all_tf, y_all, WINDOW, STEP)
print("Sequences:", X_seq.shape, "Ends:", end_idx.shape)

# ===== シーケンス単位の層化分割（train/dev/test = 60/20/20） =====
from sklearn.model_selection import train_test_split
X_tr_seq, X_tmp_seq, y_tr_seq, y_tmp_seq, end_tr, end_tmp = train_test_split(
    X_seq, y_seq, end_idx, test_size=0.4, random_state=SEED, stratify=y_seq, shuffle=True
)
X_dv_seq, X_te_seq, y_dv_seq, y_te_seq, end_dv, end_te = train_test_split(
    X_tmp_seq, y_tmp_seq, end_tmp, test_size=0.5, random_state=SEED, stratify=y_tmp_seq, shuffle=True
)
print("Seq splits:", X_tr_seq.shape, X_dv_seq.shape, X_te_seq.shape)

# ===== RF/HGB も「窓終端の行」だけで比較（同じサンプル集合） =====
X_end = X_all_tf  # 時間特徴を付与済みの行列
Xrf_tr = X_end[end_tr]
Xrf_dv = X_end[end_dv]
Xrf_te = X_end[end_te]
yrf_tr = y_all[end_tr]
yrf_dv = y_all[end_dv]
yrf_te = y_all[end_te]

# ===== 学習系列をクラスバランス =====
def balance_sequences(Xs, ys, max_per_class=None, seed=SEED):
    rng = np.random.default_rng(seed)
    uniq, cnt = np.unique(ys, return_counts=True)
    k = cnt.min()
    if max_per_class is not None:
        k = min(k, max_per_class)
    idxs = []
    for c in uniq:
        cand = np.where(ys == c)[0]
        sel = rng.choice(cand, k, replace=False)
        idxs.append(sel)
    sel = np.concatenate(idxs); rng.shuffle(sel)
    return Xs[sel], ys[sel]

X_tr_bal, y_tr_bal = balance_sequences(X_tr_seq, y_tr_seq, max_per_class=8000)

# ===== 指標 =====
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
def eval_multi(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    pw, rw, fw, _ = precision_recall_fscore_support(y_true, y_pred, average="weighted", zero_division=0)
    pm, rm, fm, _ = precision_recall_fscore_support(y_true, y_pred, average="macro", zero_division=0)
    return {"accuracy": acc, "precision_weighted": pw, "recall_weighted": rw, "f1_weighted": fw,
            "precision_macro": pm, "recall_macro": rm, "f1_macro": fm}

# ===== RF / HGB =====
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
rf = RandomForestClassifier(n_estimators=500, class_weight="balanced", random_state=SEED, n_jobs=-1)
rf.fit(Xrf_tr, yrf_tr)
y_rf = rf.predict(Xrf_te)

hgb = HistGradientBoostingClassifier(random_state=SEED)
hgb.fit(Xrf_tr, yrf_tr)
y_hgb = hgb.predict(Xrf_te)

# ===== LSTM / TCN =====
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
tf.random.set_seed(SEED)

INPUT_SHAPE = (X_tr_bal.shape[1], X_tr_bal.shape[2])

def build_lstm(input_shape, n_classes):
    m = keras.Sequential([
        layers.Input(shape=input_shape),
        layers.Masking(mask_value=0.0),
        layers.LSTM(64, return_sequences=True, dropout=0.2),
        layers.LSTM(64, dropout=0.2),
        layers.Dense(64, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(n_classes, activation="softmax")
    ])
    m.compile(optimizer=keras.optimizers.Adam(1e-3),
              loss=keras.losses.SparseCategoricalCrossentropy(),
              metrics=["accuracy"])
    return m

def build_tcn(input_shape, n_classes):
    inp = layers.Input(shape=input_shape)
    x = inp
    for d in [1,2,4,8,16]:
        res = x
        x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
        x = layers.BatchNormalization()(x); x = layers.ReLU()(x)
        x = layers.Dropout(0.2)(x)
        x = layers.Conv1D(64, 3, padding="causal", dilation_rate=d)(x)
        x = layers.BatchNormalization()(x)
        if res.shape[-1] != x.shape[-1]:
            res = layers.Conv1D(64, 1, padding="same")(res)
        x = layers.Add()([x, res]); x = layers.ReLU()(x)
    x = layers.LayerNormalization()(x)
    x = layers.GlobalAveragePooling1D()(x)
    x = layers.Dense(64, activation="relu")(x)
    x = layers.Dropout(0.3)(x)
    out = layers.Dense(n_classes, activation="softmax")(x)
    m = keras.Model(inp, out)
    m.compile(optimizer=keras.optimizers.Adam(1e-3),
              loss=keras.losses.SparseCategoricalCrossentropy(),
              metrics=["accuracy"])
    return m

def fit_nn(model, Xtr, ytr, Xdv, ydv):
    cbs = [
        keras.callbacks.ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=2, min_lr=1e-5, verbose=1),
        keras.callbacks.EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True, verbose=1),
    ]
    return model.fit(Xtr, ytr, validation_data=(Xdv, ydv),
                     epochs=EPOCHS, batch_size=BATCH, verbose=1)

lstm = build_lstm(INPUT_SHAPE, N_CLASSES)
_ = fit_nn(lstm, X_tr_bal, y_tr_bal, X_dv_seq, y_dv_seq)
y_lstm = lstm.predict(X_te_seq, verbose=0).argmax(axis=1)

tcn = build_tcn(INPUT_SHAPE, N_CLASSES)
_ = fit_nn(tcn, X_tr_bal, y_tr_bal, X_dv_seq, y_dv_seq)
y_tcn = tcn.predict(X_te_seq, verbose=0).argmax(axis=1)

# ===== Results =====
res = []
res.append({"model":"RandomForest",         **eval_multi(yrf_te, y_rf)})
res.append({"model":"SklearnHGB(XGB-like)", **eval_multi(yrf_te, y_hgb)})
res.append({"model":"LSTM",                 **eval_multi(y_te_seq, y_lstm)})
res.append({"model":"TCN",                  **eval_multi(y_te_seq, y_tcn)})

res_df = pd.DataFrame(res)
display(res_df)

# ===== Confusion matrices (present classes only) =====
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
def plot_cm(y_true, y_pred, class_names, title):
    labs = np.unique(np.concatenate([np.unique(y_true), np.unique(y_pred)]))
    cm = confusion_matrix(y_true, y_pred, labels=labs)
    fig, ax = plt.subplots(figsize=(6,5))
    ConfusionMatrixDisplay(cm, display_labels=[class_names[i] for i in labs])\
        .plot(ax=ax, cmap="Blues", colorbar=False)
    ax.set_title(title); plt.tight_layout(); plt.show()
    print(classification_report(y_true, y_pred, labels=labs,
                                target_names=[class_names[i] for i in labs],
                                zero_division=0))

plot_cm(yrf_te,  y_rf,   CLASSES, "CM - RandomForest (sequence endpoints)")
plot_cm(yrf_te,  y_hgb,  CLASSES, "CM - HGB (sequence endpoints)")
plot_cm(y_te_seq, y_lstm, CLASSES, "CM - LSTM")
plot_cm(y_te_seq, y_tcn,  CLASSES, "CM - TCN")
```

```python
# ============================================
# Room Position (Multiclass) — Tree-only (RF/HGB)
# Robust split (day/hour/pseudo + embargo), tolerant to missing classes in splits
# Confusion matrices & reports always over the full label set (no crash)
# ============================================

# ===== User Config =====
DATA_FILE   = "0919Living_labeled.csv"
TARGET      = "label_room"
TIMESTAMP   = "timestamp"    # 並び替え/ブロック化に使用
EXCLUDE     = {TARGET, TIMESTAMP, "Occupied", "Num of People"}  # 不要列があれば追加
SEED        = 42
TRAIN_RATIO = 0.60
DEV_RATIO   = 0.20
EMB_ROWS    = 60            # 初期エンバーゴ行数（自動縮小あり）
MIN_ROWS    = 200           # 各 split の最低必要行数（足りなければ縮小/フォールバック）
OUT_DIR     = "tree_only_outputs_adaptive"

# ===== Imports =====
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (accuracy_score, precision_recall_fscore_support,
                             confusion_matrix, classification_report, ConfusionMatrixDisplay)
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier

np.random.seed(SEED); Path(OUT_DIR).mkdir(exist_ok=True, parents=True)

# ===== Load =====
df = pd.read_csv(DATA_FILE, low_memory=False)
if TIMESTAMP in df.columns:
    df[TIMESTAMP] = pd.to_datetime(df[TIMESTAMP], errors="coerce")
    df = df.sort_values(TIMESTAMP).reset_index(drop=True)
df = df.dropna(subset=[TARGET]).copy()

# ===== Target =====
le = LabelEncoder()
y_all = le.fit_transform(df[TARGET].astype(str))
CLASSES = le.classes_.tolist()
NCLASS = len(CLASSES)
print("Classes:", CLASSES, "N:", len(df))

# ===== Features =====
feat_cols = [c for c in df.columns if c not in EXCLUDE]
X_df = df[feat_cols].copy()
cat_cols = [c for c in X_df.columns if X_df[c].dtype == "object"]
num_cols = [c for c in X_df.columns if c not in cat_cols]

pre = ColumnTransformer([
    ("num", Pipeline([("imp", SimpleImputer(strategy="median")),
                      ("sc",  StandardScaler())]), num_cols),
    ("cat", Pipeline([("imp", SimpleImputer(strategy="most_frequent")),
                      ("oh",  OneHotEncoder(handle_unknown="ignore"))]), cat_cols),
])

# ===== Block helpers =====
def make_blocks(series_ts, mode="day"):
    """mode: 'day' | 'hour' | 'pseudo'"""
    if series_ts is None:
        return None
    ts_series = pd.Series(series_ts)  # DatetimeIndex対応
    if mode == "day":
        return ts_series.dt.date.astype(str).values
    if mode == "hour":
        return ts_series.dt.strftime("%Y-%m-%d %H").values
    if mode == "pseudo":
        n = len(series_ts); nb = 30
        bins = np.linspace(0, n-1, nb+1, dtype=int)
        blk = np.zeros(n, dtype=int)
        for b in range(nb):
            blk[bins[b]:bins[b+1]] = b
        return blk.astype(str)
    return None

def chronological_order(block_ids):
    uniq, first_idx = np.unique(block_ids, return_index=True)
    return uniq[np.argsort(first_idx)]

def build_masks(block_ids, train_ratio, dev_ratio, emb_rows):
    order = chronological_order(block_ids)
    nB = len(order)
    n_tr = max(1, int(nB*train_ratio))
    n_dv = max(1 if nB>=3 else 0, int(nB*dev_ratio))
    blk_tr = order[:n_tr]; blk_dv = order[n_tr:n_tr+n_dv]; blk_te = order[n_tr+n_dv:]
    # 行範囲
    block_to_range = {}
    for b in order:
        idxs = np.where(block_ids == b)[0]
        block_to_range[b] = (int(idxs.min()), int(idxs.max()))
    def mask_blocks(keep_blocks, drop_near_blocks, emb):
        keep = np.isin(block_ids, keep_blocks)
        if len(drop_near_blocks) == 0: return keep
        ban = np.zeros(len(block_ids), dtype=bool)
        for b in np.unique(drop_near_blocks):
            a, z = block_to_range[b]
            lo, hi = max(0, a-emb), min(len(block_ids)-1, z+emb)
            ban[lo:hi+1] = True
        return keep & (~ban)
    m_tr = mask_blocks(blk_tr, np.concatenate([blk_dv, blk_te]) if len(blk_dv)+len(blk_te)>0 else [], emb_rows)
    m_dv = mask_blocks(blk_dv, np.concatenate([blk_tr, blk_te]) if len(blk_tr)+len(blk_te)>0 else [], emb_rows)
    m_te = mask_blocks(blk_te, np.concatenate([blk_tr, blk_dv]) if len(blk_tr)+len(blk_dv)>0 else [], emb_rows)
    return m_tr, m_dv, m_te, order

def check_minimum(idx_tr, idx_dv, idx_te, min_rows=MIN_ROWS):
    # クラス完全包含は必須にしない（現実的な評価を優先）
    return (len(idx_tr) >= min_rows) and (len(idx_dv) >= min_rows) and (len(idx_te) >= min_rows)

# ===== Choose split with fallback =====
ts_vals = df[TIMESTAMP].values if TIMESTAMP in df.columns else None
candidates = [("day", EMB_ROWS), ("day", EMB_ROWS//2), ("day", 0),
              ("hour", EMB_ROWS), ("hour", EMB_ROWS//2), ("hour", 0),
              ("pseudo", EMB_ROWS), ("pseudo", EMB_ROWS//2), ("pseudo", 0)]
chosen = None
for mode, emb in candidates:
    blk_ids = make_blocks(ts_vals, mode=mode)
    if blk_ids is None: continue
    m_tr, m_dv, m_te, order = build_masks(blk_ids, TRAIN_RATIO, DEV_RATIO, emb)
    idx_tr = np.where(m_tr)[0]; idx_dv = np.where(m_dv)[0]; idx_te = np.where(m_te)[0]
    print(f"[try] mode={mode}, embargo={emb} -> sizes train={len(idx_tr)}, dev={len(idx_dv)}, test={len(idx_te)}")
    if check_minimum(idx_tr, idx_dv, idx_te, MIN_ROWS):
        chosen = (mode, emb, idx_tr, idx_dv, idx_te)
        print("  -> accepted")
        break

if chosen is None:
    # 最後の手段：単純時系列分割（リークの可能性あり）
    n = len(df); i1 = int(n*TRAIN_RATIO); i2 = int(n*(TRAIN_RATIO+DEV_RATIO))
    idx_tr = np.arange(0, i1); idx_dv = np.arange(i1, i2); idx_te = np.arange(i2, n)
    print("WARNING: fell back to naive chronological split (no embargo). "
          "Results may be optimistic if adjacent leakage exists.")
else:
    mode, emb, idx_tr, idx_dv, idx_te = chosen
    print(f"Chosen split: mode={mode}, embargo={emb}")

print(f"Final split sizes: train={len(idx_tr)}, dev={len(idx_dv)}, test={len(idx_te)}")

# ===== Fit preprocess on TRAIN only =====
X_tr = pre.fit_transform(X_df.iloc[idx_tr])
X_dv = pre.transform(X_df.iloc[idx_dv])
X_te = pre.transform(X_df.iloc[idx_te])
if hasattr(X_tr, "toarray"):
    X_tr = X_tr.toarray(); X_dv = X_dv.toarray(); X_te = X_te.toarray()
y_tr, y_dv, y_te = y_all[idx_tr], y_all[idx_dv], y_all[idx_te]

# ===== Class distribution plots =====
def plot_class_dist(y, classes, title, path):
    vals, cnt = np.unique(y, return_counts=True)
    labs = [classes[v] for v in vals]
    plt.figure(figsize=(6,3))
    plt.bar(labs, cnt)
    plt.title(title); plt.ylabel("count"); plt.xticks(rotation=20)
    plt.tight_layout(); plt.savefig(path, dpi=150); plt.close()

plot_class_dist(y_tr, CLASSES, "Train class distribution", f"{OUT_DIR}/dist_train.png")
plot_class_dist(y_te, CLASSES, "Test class distribution",  f"{OUT_DIR}/dist_test.png")

# ===== Metrics =====
def eval_multi(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    pw, rw, fw, _ = precision_recall_fscore_support(y_true, y_pred, average="weighted", zero_division=0)
    pm, rm, fm, _ = precision_recall_fscore_support(y_true, y_pred, average="macro", zero_division=0)
    return {"accuracy": acc,
            "precision_weighted": pw, "recall_weighted": rw, "f1_weighted": fw,
            "precision_macro": pm, "recall_macro": rm, "f1_macro": fm}

# ===== Train & Evaluate =====
rf = RandomForestClassifier(n_estimators=600, class_weight="balanced",
                            random_state=SEED, n_jobs=-1)
rf.fit(X_tr, y_tr)
y_rf = rf.predict(X_te)
res_rf = {"model":"RandomForest", **eval_multi(y_te, y_rf)}

hgb = HistGradientBoostingClassifier(random_state=SEED)
hgb.fit(X_tr, y_tr)
y_hgb = hgb.predict(X_te)
res_hgb = {"model":"SklearnHGB(XGB-like)", **eval_multi(y_te, y_hgb)}

# ===== Save CM & Reports — ALWAYS over full label set =====
def save_cm_and_report_full(y_true, y_pred, classes, title, stem):
    labels_full = np.arange(len(classes))  # 0..K-1（全クラス）
    cm = confusion_matrix(y_true, y_pred, labels=labels_full)
    fig, ax = plt.subplots(figsize=(5.8,4.6))
    ConfusionMatrixDisplay(cm, display_labels=classes).plot(ax=ax, cmap="Blues", colorbar=False)
    ax.set_title(title)
    plt.tight_layout(); plt.savefig(f"{OUT_DIR}/cm_{stem}.png", dpi=150); plt.close()
    # 一部クラスが欠けていても OK
    with open(f"{OUT_DIR}/report_{stem}.txt","w") as f:
        f.write(classification_report(
            y_true, y_pred, labels=labels_full, target_names=classes, zero_division=0
        ))

# 欠けクラスの警告（議論材料として）
def warn_missing_classes(y, where):
    present = np.unique(y)
    missing = [c for i,c in enumerate(CLASSES) if i not in present]
    if missing:
        print(f"[WARN] {where} split is missing classes: {missing}")

warn_missing_classes(y_tr, "TRAIN"); warn_missing_classes(y_dv, "DEV"); warn_missing_classes(y_te, "TEST")

save_cm_and_report_full(y_te, y_rf,  CLASSES, "Confusion Matrix - RF (All labels)",  "rf")
save_cm_and_report_full(y_te, y_hgb, CLASSES, "Confusion Matrix - HGB (All labels)", "hgb")

# ===== Results table =====
res_df = pd.DataFrame([res_rf, res_hgb])
res_df.to_csv(f"{OUT_DIR}/metrics_tree_only_adaptive.csv", index=False)

def to_percent(x): return f"{100*x:.4f}%"
show = res_df.copy()
for c in ["accuracy","precision_weighted","recall_weighted","f1_weighted",
          "precision_macro","recall_macro","f1_macro"]:
    show[c] = show[c].map(to_percent)

print("\n=== Tree-only Results (percent) ===")
print(show)
print(f"\nSaved to: {OUT_DIR}")
```

```python
# ============================================
# Multiclass Room Position — Tree-only Robust Grouped CV
# Models: RandomForest, HistGradientBoosting (XGB-like)
# CV: GroupShuffleSplit (group = hour), with retries to ensure non-empty, class coverage
# Metrics: Accuracy / Weighted-F1 / Macro-F1
# Outputs: per-fold CM & reports, mean table (percent, 4 sig figs)
# ============================================

# ===== User Config =====
DATA_FILE = "0919Living_labeled.csv"
TARGET    = "label_room"
TIMESTAMP = "timestamp"
EXCLUDE   = {TARGET, TIMESTAMP, "Occupied", "Num of People"}  # 必要に応じて追加
N_FOLDS   = 5
SEED      = 42
TEST_SIZE = 0.20     # 各リピートのテスト割合（グループ単位）
MIN_TEST  = 300      # テストの最小行数（空fold防止）
RETRY_MAX = 60       # クラス欠損が大きい場合のリトライ上限
OUT_DIR   = "tree_only_group_shuffle_cv"

# ===== Imports =====
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (accuracy_score, precision_recall_fscore_support,
                             confusion_matrix, classification_report, ConfusionMatrixDisplay)
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.model_selection import GroupShuffleSplit

np.random.seed(SEED)
Path(OUT_DIR).mkdir(exist_ok=True, parents=True)

# ===== Load =====
df = pd.read_csv(DATA_FILE, low_memory=False)
assert TARGET in df.columns, f"'{TARGET}' 列が見つかりません。"
assert TIMESTAMP in df.columns, f"'{TIMESTAMP}' 列が見つかりません。"
df[TIMESTAMP] = pd.to_datetime(df[TIMESTAMP], errors="coerce")
df = df.dropna(subset=[TARGET, TIMESTAMP]).sort_values(TIMESTAMP).reset_index(drop=True)

# group = 時間（YYYY-mm-dd HH）
groups = df[TIMESTAMP].dt.strftime("%Y-%m-%d %H").values

# ===== Target =====
le = LabelEncoder()
y_all = le.fit_transform(df[TARGET].astype(str))
CLASSES = le.classes_.tolist(); NCLASS = len(CLASSES)
print("Classes:", CLASSES, "N:", len(df))

# ===== Features =====
feat_cols = [c for c in df.columns if c not in EXCLUDE]
X_df = df[feat_cols].copy()
cat_cols = [c for c in X_df.columns if X_df[c].dtype == "object"]
num_cols = [c for c in X_df.columns if c not in cat_cols]

pre = ColumnTransformer([
    ("num", Pipeline([("imp", SimpleImputer(strategy="median")),
                      ("sc",  StandardScaler())]), num_cols),
    ("cat", Pipeline([("imp", SimpleImputer(strategy="most_frequent")),
                      ("oh",  OneHotEncoder(handle_unknown="ignore"))]), cat_cols),
])

# ===== Robust group splits with retries =====
def iter_group_shuffles(y, groups, n_folds=N_FOLDS, test_size=TEST_SIZE,
                        min_test=MIN_TEST, retry_max=RETRY_MAX, seed=SEED):
    rng = np.random.RandomState(seed)
    uniq_groups = np.unique(groups)
    for k in range(n_folds):
        ok = False
        for attempt in range(retry_max):
            gss = GroupShuffleSplit(n_splits=1, test_size=test_size,
                                    random_state=rng.randint(0, 10**9))
            (tr_idx, te_idx), = gss.split(np.zeros(len(y)), y, groups)
            # 必須条件: テストが十分な行数
            if len(te_idx) < min_test:
                continue
            # 望ましい条件: テストに最低2クラス、できれば全クラス
            te_classes = np.unique(y[te_idx])
            if len(te_classes) >= min(3, NCLASS):  # 3クラス以上を目標（全クラスでなくてもOK）
                ok = True
                break
        if not ok:
            print(f"[Fold {k+1}] WARN: fell back to size-only condition "
                  f"(test={len(te_idx)}, classes={list(le.inverse_transform(np.unique(y[te_idx])))}).")
        yield tr_idx, te_idx

# ===== Models =====
MODELS = {
    "RandomForest": RandomForestClassifier(
        n_estimators=600, class_weight="balanced",
        random_state=SEED, n_jobs=-1
    ),
    "SklearnHGB(XGB-like)": HistGradientBoostingClassifier(
        random_state=SEED
    )
}

# ===== Metrics helpers =====
def eval_multi(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    pw, rw, fw, _ = precision_recall_fscore_support(y_true, y_pred, average="weighted", zero_division=0)
    pm, rm, fm, _ = precision_recall_fscore_support(y_true, y_pred, average="macro", zero_division=0)
    return {"accuracy": acc,
            "precision_weighted": pw, "recall_weighted": rw, "f1_weighted": fw,
            "precision_macro": pm, "recall_macro": rm, "f1_macro": fm}

def save_cm_and_report_full(y_true, y_pred, classes, title, stem):
    labels_full = np.arange(len(classes))
    cm = confusion_matrix(y_true, y_pred, labels=labels_full)
    fig, ax = plt.subplots(figsize=(5.8,4.6))
    ConfusionMatrixDisplay(cm, display_labels=classes).plot(ax=ax, cmap="Blues", colorbar=False, values_format="d")
    ax.set_title(title); plt.tight_layout(); plt.savefig(f"{OUT_DIR}/cm_{stem}.png", dpi=150); plt.close()
    with open(f"{OUT_DIR}/report_{stem}.txt","w") as f:
        f.write(classification_report(y_true, y_pred,
                                      labels=labels_full, target_names=classes, zero_division=0))

def pct(x): return f"{100*x:.4f}%"

# ===== CV loop =====
all_rows = []
for fold_id, (tr_idx, te_idx) in enumerate(iter_group_shuffles(y_all, groups), start=1):
    # Safety: skip empty test fold（念のため）
    if len(te_idx) == 0:
        print(f"[Fold {fold_id}] SKIP (empty test)")
        continue

    # 参考: テストに欠けているクラスを表示
    miss = [CLASSES[i] for i in range(NCLASS) if i not in np.unique(y_all[te_idx])]
    if miss:
        print(f"[Fold {fold_id}] test missing classes:", miss)

    # 前処理は train のみで fit
    X_tr = pre.fit_transform(X_df.iloc[tr_idx])
    X_te = pre.transform(X_df.iloc[te_idx])
    if hasattr(X_tr, "toarray"):
        X_tr = X_tr.toarray(); X_te = X_te.toarray()
    y_tr, y_te = y_all[tr_idx], y_all[te_idx]

    for name, model in MODELS.items():
        clf = model
        clf.fit(X_tr, y_tr)
        y_hat = clf.predict(X_te)
        m = eval_multi(y_te, y_hat)
        all_rows.append({"fold": fold_id, "model": name, **m})
        save_cm_and_report_full(y_te, y_hat, CLASSES,
                                f"Confusion Matrix - {name} (fold {fold_id})",
                                f"{name.replace('(','_').replace(')','').replace(' ','_')}_f{fold_id}")

# ===== 集計 =====
res_df = pd.DataFrame(all_rows)
res_df.to_csv(f"{OUT_DIR}/cv_raw_metrics.csv", index=False)

mean_df = (res_df.groupby("model")[["accuracy","precision_weighted","recall_weighted",
                                    "f1_weighted","precision_macro","recall_macro","f1_macro"]]
           .mean().reset_index())

show = mean_df.copy()
for c in ["accuracy","precision_weighted","recall_weighted","f1_weighted",
          "precision_macro","recall_macro","f1_macro"]:
    show[c] = show[c].map(pct)

show.to_csv(f"{OUT_DIR}/cv_summary_percent.csv", index=False)
print("\n=== CV Summary (mean over folds, percent) ===")
print(show)
print(f"\nArtifacts saved in: {OUT_DIR}")
```

```python
# ============================================
# Multiclass Room Position — Tree-only CV (Robust, Class-covered test folds)
#   - Models: RandomForest, HistGradientBoosting (XGB-like)
#   - Group = hour ("YYYY-mm-dd HH"), Test fold MUST cover ALL classes
#   - Embargo around test groups to reduce leakage
#   - Metrics: Accuracy / Weighted-F1 / Macro-F1
# ============================================

# ===== User Config =====
DATA_FILE = "0919Living_labeled.csv"
TARGET    = "label_room"
TIMESTAMP = "timestamp"
EXCLUDE   = {TARGET, TIMESTAMP, "Occupied", "Num of People"}  # 必要なら追加
N_FOLDS   = 5
SEED      = 42
TEST_GROUP_RATIO = 0.20   # テストに使う“時間グループ”の割合（目安）
EMB_ROWS  = 60            # テスト区間の前後に除外する行数（リーク低減）
OUT_DIR   = "tree_only_group_cover_cv"

# ===== Imports =====
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (accuracy_score, precision_recall_fscore_support,
                             confusion_matrix, classification_report, ConfusionMatrixDisplay)
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier

np.random.seed(SEED)
Path(OUT_DIR).mkdir(exist_ok=True, parents=True)

# ===== Load =====
df = pd.read_csv(DATA_FILE, low_memory=False)
assert TARGET in df.columns and TIMESTAMP in df.columns
df[TIMESTAMP] = pd.to_datetime(df[TIMESTAMP], errors="coerce")
df = df.dropna(subset=[TARGET, TIMESTAMP]).sort_values(TIMESTAMP).reset_index(drop=True)

# Group = hour
group_ids = df[TIMESTAMP].dt.strftime("%Y-%m-%d %H").values
groups_unique, group_first_idx = np.unique(group_ids, return_index=True)
groups_unique = groups_unique[np.argsort(group_first_idx)]  # 時系列順ユニーク

# ===== Target =====
le = LabelEncoder()
y_all = le.fit_transform(df[TARGET].astype(str))
CLASSES = le.classes_.tolist(); K = len(CLASSES)
print("Classes:", CLASSES, "N:", len(df))

# ===== Features =====
feat_cols = [c for c in df.columns if c not in EXCLUDE]
X_df = df[feat_cols].copy()
cat_cols = [c for c in X_df.columns if X_df[c].dtype == "object"]
num_cols = [c for c in X_df.columns if c not in cat_cols]

pre = ColumnTransformer([
    ("num", Pipeline([("imp", SimpleImputer(strategy="median")),
                      ("sc",  StandardScaler())]), num_cols),
    ("cat", Pipeline([("imp", SimpleImputer(strategy="most_frequent")),
                      ("oh",  OneHotEncoder(handle_unknown="ignore"))]), cat_cols),
])

# ===== Helper: indices belonging to a set of groups =====
def idx_of_groups(selected_groups):
    mask = np.isin(group_ids, selected_groups)
    return np.where(mask)[0]

# ===== Build folds: test covers ALL classes (greedy over groups) =====
def build_stratified_group_folds(n_folds=N_FOLDS, seed=SEED, test_ratio=TEST_GROUP_RATIO, max_trials=2000):
    rng = np.random.RandomState(seed)
    folds = []
    # 事前に各グループのクラス分布を計算
    group_to_classes = {}
    for g in groups_unique:
        idx = idx_of_groups([g])
        group_to_classes[g] = set(np.unique(y_all[idx]).tolist())

    total_groups = len(groups_unique)
    target_test_groups = max(1, int(total_groups * test_ratio))

    # 反復しながら、毎回ランダム順で“全クラス被覆”までグループを貪欲に足す
    for f in range(n_folds):
        best = None
        for _ in range(max_trials):
            order = groups_unique.copy()
            rng.shuffle(order)
            covered = set()
            chosen = []
            for g in order:
                chosen.append(g)
                covered |= group_to_classes[g]
                if (len(covered) == K) and (len(chosen) >= target_test_groups//2):
                    # 最低限のサイズも確保、被覆達成で早期終了
                    break
            # サイズ調整（大き過ぎる/小さ過ぎる場合は再試行）
            te_idx = idx_of_groups(chosen)
            if len(te_idx) < 500:   # テスト最低行数の目安
                continue
            if len(te_idx) > len(df)*0.35:  # テストが大き過ぎるのは避ける
                continue
            if len(covered) < K:
                continue
            best = chosen
            break

        if best is None:
            # 妥協：後半の連続グループで被覆達成まで伸ばす
            covered = set()
            chosen = []
            for g in groups_unique[::-1]:
                chosen.append(g)
                covered |= group_to_classes[g]
                if len(covered) == K:
                    break
            best = chosen

        test_groups = np.array(best)
        test_idx = idx_of_groups(test_groups)
        train_idx = np.setdiff1d(np.arange(len(df)), test_idx)
        folds.append((train_idx, test_idx, test_groups))
    return folds

# ===== Embargo: remove rows around test segments from train =====
def apply_embargo_masks(train_idx, test_idx, emb_rows=EMB_ROWS):
    # 連続区間ごとにテストの前後 emb_rows を train から削除
    train_mask = np.zeros(len(df), dtype=bool); train_mask[train_idx] = True
    test_mask  = np.zeros(len(df), dtype=bool); test_mask[test_idx]  = True
    # テストの連続ブロック抽出
    where = np.where(test_mask)[0]
    if len(where):
        splits = np.where(np.diff(where) > 1)[0]
        starts = np.r_[where[0], where[splits+1]]
        ends   = np.r_[where[splits], where[-1]]
        for s,e in zip(starts, ends):
            lo = max(0, s - emb_rows)
            hi = min(len(df)-1, e + emb_rows)
            train_mask[lo:hi+1] = False
    new_train_idx = np.where(train_mask)[0]
    return new_train_idx, test_idx

# ===== Models =====
MODELS = {
    "RandomForest": RandomForestClassifier(
        n_estimators=600, class_weight="balanced", random_state=SEED, n_jobs=-1
    ),
    "SklearnHGB(XGB-like)": HistGradientBoostingClassifier(random_state=SEED)
}

# ===== Metrics =====
def eval_multi(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    pw, rw, fw, _ = precision_recall_fscore_support(y_true, y_pred, average="weighted", zero_division=0)
    pm, rm, fm, _ = precision_recall_fscore_support(y_true, y_pred, average="macro", zero_division=0)
    return {"accuracy": acc,
            "precision_weighted": pw, "recall_weighted": rw, "f1_weighted": fw,
            "precision_macro": pm, "recall_macro": rm, "f1_macro": fm}

def save_cm_and_report_full(y_true, y_pred, classes, title, stem):
    labels_full = np.arange(len(classes))
    cm = confusion_matrix(y_true, y_pred, labels=labels_full)
    fig, ax = plt.subplots(figsize=(5.8,4.6))
    ConfusionMatrixDisplay(cm, display_labels=classes).plot(ax=ax, cmap="Blues", colorbar=False, values_format="d")
    ax.set_title(title); plt.tight_layout(); plt.savefig(f"{OUT_DIR}/cm_{stem}.png", dpi=150); plt.close()
    with open(f"{OUT_DIR}/report_{stem}.txt","w") as f:
        f.write(classification_report(y_true, y_pred,
                                      labels=labels_full, target_names=classes, zero_division=0))

def pct(x): return f"{100*x:.4f}%"

# ===== Build folds and run CV =====
fold_specs = build_stratified_group_folds(n_folds=N_FOLDS, seed=SEED, test_ratio=TEST_GROUP_RATIO)
all_rows = []

for fold_id, (tr_idx0, te_idx0, te_groups) in enumerate(fold_specs, start=1):
    # embargo 適用
    tr_idx, te_idx = apply_embargo_masks(tr_idx0, te_idx0, emb_rows=EMB_ROWS)

    # テストクラスの被覆確認（議論材料として出力）
    te_classes = list(le.inverse_transform(np.unique(y_all[te_idx])))
    missing = [c for c in CLASSES if c not in te_classes]
    if missing:
        print(f"[Fold {fold_id}] WARN test missing classes after embargo:", missing)

    # 前処理は train のみで fit
    X_tr = pre.fit_transform(X_df.iloc[tr_idx])
    X_te = pre.transform(X_df.iloc[te_idx])
    if hasattr(X_tr, "toarray"): X_tr = X_tr.toarray(); X_te = X_te.toarray()
    y_tr, y_te = y_all[tr_idx], y_all[te_idx]

    for name, model in MODELS.items():
        clf = model
        clf.fit(X_tr, y_tr)
        y_hat = clf.predict(X_te)
        m = eval_multi(y_te, y_hat)
        all_rows.append({"fold": fold_id, "model": name, **m})
        save_cm_and_report_full(y_te, y_hat, CLASSES,
                                f"Confusion Matrix - {name} (fold {fold_id})",
                                f"{name.replace('(','_').replace(')','').replace(' ','_')}_f{fold_id}")

# ===== 集計 =====
res_df = pd.DataFrame(all_rows)
res_df.to_csv(f"{OUT_DIR}/cv_raw_metrics.csv", index=False)
mean_df = (res_df.groupby("model")[["accuracy","precision_weighted","recall_weighted",
                                    "f1_weighted","precision_macro","recall_macro","f1_macro"]]
           .mean().reset_index())

show = mean_df.copy()
for c in ["accuracy","precision_weighted","recall_weighted","f1_weighted",
          "precision_macro","recall_macro","f1_macro"]:
    show[c] = show[c].map(pct)

show.to_csv(f"{OUT_DIR}/cv_summary_percent.csv", index=False)
print("\n=== CV Summary (mean over folds, percent) ===")
print(show)
print(f"\nArtifacts saved in: {OUT_DIR}")
```

```python
# ============================================
# 出力一式: 表（%・有効数字4桁）、混同行列、寄与センサ（PI）、PR/ROC + macro AP/AUC
# データ: 0919Living_labeled.csv
# モデル: RandomForest, HistGradientBoosting (XGB-like)
# 分割: 時間グループ（hour）でテストに全クラスを被覆、エンバーゴあり
# scikit-learn >= 1.2 互換（OneHotEncoder: sparse_output=False）
# ============================================

# ===== ユーザ設定 =====
DATA_FILE = "0919Living_labeled.csv"
TARGET    = "label_room"
TIMESTAMP = "timestamp"
EXCLUDE   = {TARGET, TIMESTAMP, "Occupied", "Num of People"}  # 必要に応じて追加
N_FOLDS   = 5
SEED      = 42
TEST_GROUP_RATIO = 0.20   # テストに使う時間グループの割合（目安）
EMB_ROWS  = 60            # テスト区間の前後を学習から除外（リーク抑制）
OUT_DIR   = "final_artifacts_treecv"

# ===== import =====
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler, label_binarize
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (accuracy_score, precision_recall_fscore_support,
                             confusion_matrix, classification_report, ConfusionMatrixDisplay,
                             PrecisionRecallDisplay, RocCurveDisplay,
                             average_precision_score, roc_auc_score)
from sklearn.ensemble import RandomForestClassifier, HistGradientBoostingClassifier
from sklearn.inspection import permutation_importance

np.random.seed(SEED)
Path(OUT_DIR).mkdir(exist_ok=True, parents=True)

# ===== データ読み込み =====
df = pd.read_csv(DATA_FILE, low_memory=False)
assert TARGET in df.columns and TIMESTAMP in df.columns
df[TIMESTAMP] = pd.to_datetime(df[TIMESTAMP], errors="coerce")
df = df.dropna(subset=[TARGET, TIMESTAMP]).sort_values(TIMESTAMP).reset_index(drop=True)

# 時間グループ（1時間単位）
group_ids = df[TIMESTAMP].dt.strftime("%Y-%m-%d %H").values
groups_unique, group_first_idx = np.unique(group_ids, return_index=True)
groups_unique = groups_unique[np.argsort(group_first_idx)]  # 時系列順でユニーク化

# ===== 目的変数 =====
le = LabelEncoder()
y_all = le.fit_transform(df[TARGET].astype(str))
CLASSES = le.classes_.tolist(); K = len(CLASSES)
print("Classes:", CLASSES, "N:", len(df))

# ===== 特徴量 =====
feat_cols = [c for c in df.columns if c not in EXCLUDE]
X_df = df[feat_cols].copy()

# 恒常NaN列は事前にドロップ（警告解消＆高速化）
all_nan_cols = [c for c in X_df.columns if not X_df[c].notna().any()]
if all_nan_cols:
    print("[INFO] Drop all-NaN columns:", all_nan_cols)
    X_df = X_df.drop(columns=all_nan_cols)
    feat_cols = [c for c in feat_cols if c not in all_nan_cols]

cat_cols = [c for c in X_df.columns if X_df[c].dtype == "object"]
num_cols = [c for c in X_df.columns if c not in cat_cols]

# --- OneHotEncoder: sklearn>=1.2 は sparse_output, <1.2 は sparse ---
from packaging import version
import sklearn
if version.parse(sklearn.__version__) >= version.parse("1.2"):
    ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
else:
    ohe = OneHotEncoder(handle_unknown="ignore", sparse=False)

pre = ColumnTransformer([
    ("num", Pipeline([("imp", SimpleImputer(strategy="median")),
                      ("sc",  StandardScaler())]), num_cols),
    ("cat", Pipeline([("imp", SimpleImputer(strategy="most_frequent")),
                      ("oh",  ohe)]), cat_cols),
])

def get_feature_names(preprocessor):
    names = []
    if "num" in preprocessor.named_transformers_:
        names += list(num_cols)
    if "cat" in preprocessor.named_transformers_ and len(cat_cols) > 0:
        oh = preprocessor.named_transformers_["cat"].named_steps["oh"]
        # sklearn >=1.0 では get_feature_names_out が推奨
        try:
            cat_names = oh.get_feature_names_out(cat_cols).tolist()
        except Exception:
            # 旧APIフォールバック
            cat_names = []
            for i, c in enumerate(cat_cols):
                cats = oh.categories_[i]
                cat_names += [f"{c}={v}" for v in cats]
        names += cat_names
    return names

# ===== グループ関数 =====
def idx_of_groups(selected_groups):
    mask = np.isin(group_ids, selected_groups)
    return np.where(mask)[0]

def build_stratified_group_folds(n_folds=N_FOLDS, seed=SEED, test_ratio=TEST_GROUP_RATIO, max_trials=2000):
    """テストが全クラスを被覆するよう、グループ（hour）を貪欲選択"""
    rng = np.random.RandomState(seed)
    folds = []
    # 各グループのクラス集合
    group_to_classes = {}
    for g in groups_unique:
        idx = idx_of_groups([g])
        group_to_classes[g] = set(np.unique(y_all[idx]).tolist())

    total_groups = len(groups_unique)
    target_test_groups = max(1, int(total_groups * test_ratio))

    for f in range(n_folds):
        best = None
        for _ in range(max_trials):
            order = groups_unique.copy(); rng.shuffle(order)
            covered = set(); chosen = []
            for g in order:
                chosen.append(g)
                covered |= group_to_classes[g]
                if (len(covered) == K) and (len(chosen) >= max(1, target_test_groups//2)):
                    break
            te_idx = idx_of_groups(chosen)
            if len(te_idx) < 500:                 # テスト最小行数（調整可）
                continue
            if len(te_idx) > len(df)*0.35:        # テストが大き過ぎるのは回避
                continue
            if len(covered) < K:                  # 全クラス被覆
                continue
            best = chosen; break

        if best is None:
            # 妥協策: 末尾からクラス被覆達成まで
            covered = set(); chosen = []
            for g in groups_unique[::-1]:
                chosen.append(g)
                covered |= group_to_classes[g]
                if len(covered) == K:
                    break
            best = chosen

        test_groups = np.array(best)
        test_idx = idx_of_groups(test_groups)
        train_idx = np.setdiff1d(np.arange(len(df)), test_idx)
        folds.append((train_idx, test_idx, test_groups))
    return folds

def apply_embargo_masks(train_idx, test_idx, emb_rows=EMB_ROWS):
    """テスト区間の前後 emb_rows を学習から除外（リーク抑制）"""
    train_mask = np.zeros(len(df), dtype=bool); train_mask[train_idx] = True
    test_mask  = np.zeros(len(df), dtype=bool); test_mask[test_idx]  = True
    where = np.where(test_mask)[0]
    if len(where):
        splits = np.where(np.diff(where) > 1)[0]
        starts = np.r_[where[0], where[splits+1]]
        ends   = np.r_[where[splits], where[-1]]
        for s,e in zip(starts, ends):
            lo = max(0, s - emb_rows); hi = min(len(df)-1, e + emb_rows)
            train_mask[lo:hi+1] = False
    return np.where(train_mask)[0], test_idx

# ===== モデル =====
MODELS = {
    "RandomForest": RandomForestClassifier(
        n_estimators=600, class_weight="balanced", random_state=SEED, n_jobs=-1
    ),
    "SklearnHGB(XGB-like)": HistGradientBoostingClassifier(random_state=SEED)
}

# ===== メトリクス =====
def eval_multi(y_true, y_pred):
    acc = accuracy_score(y_true, y_pred)
    pw, rw, fw, _ = precision_recall_fscore_support(y_true, y_pred, average="weighted", zero_division=0)
    pm, rm, fm, _ = precision_recall_fscore_support(y_true, y_pred, average="macro", zero_division=0)
    return {"accuracy": acc,
            "precision_weighted": pw, "recall_weighted": rw, "f1_weighted": fw,
            "precision_macro": pm, "recall_macro": rm, "f1_macro": fm}

def class_report_df(y_true, y_pred, classes):
    rep = classification_report(y_true, y_pred, labels=np.arange(len(classes)),
                                target_names=classes, zero_division=0, output_dict=True)
    df_rep = pd.DataFrame(rep).T.reset_index().rename(columns={"index":"label"})
    return df_rep

def pct_4sf(x):
    # 有効数字4桁の%表記
    if x == 0:
        return "0.000%"
    s = f"{100*x:.6g}"
    return s + "%"

# ===== CV 実行 =====
fold_specs = build_stratified_group_folds(n_folds=N_FOLDS, seed=SEED, test_ratio=TEST_GROUP_RATIO)
all_rows, perclass_rows = [], []

for fold_id, (tr_idx0, te_idx0, te_groups) in enumerate(fold_specs, start=1):
    tr_idx, te_idx = apply_embargo_masks(tr_idx0, te_idx0, emb_rows=EMB_ROWS)

    # 前処理（trainのみfit）
    X_tr = pre.fit_transform(X_df.iloc[tr_idx]); y_tr = y_all[tr_idx]
    X_te = pre.transform(X_df.iloc[te_idx]);     y_te = y_all[te_idx]
    if hasattr(X_tr, "toarray"): X_tr = X_tr.toarray(); X_te = X_te.toarray()

    feat_names = get_feature_names(pre)

    for name, model in MODELS.items():
        clf = model
        clf.fit(X_tr, y_tr)
        y_hat = clf.predict(X_te)

        # --- (1) 表用: foldメトリクス ---
        m = eval_multi(y_te, y_hat)
        all_rows.append({"fold": fold_id, "model": name, **m})

        # --- (2) 混同行列（保存） ---
        cm = confusion_matrix(y_te, y_hat, labels=np.arange(K))
        fig, ax = plt.subplots(figsize=(6,5))
        ConfusionMatrixDisplay(cm, display_labels=CLASSES).plot(ax=ax, cmap="Blues", colorbar=False, values_format="d")
        ax.set_title(f"Confusion Matrix - {name} (fold {fold_id})")
        plt.tight_layout(); plt.savefig(f"{OUT_DIR}/cm_{name}_f{fold_id}.png", dpi=150); plt.close()

        # クラス別詳細（csv保存）
        rep_df = class_report_df(y_te, y_hat, CLASSES)
        rep_df.insert(0, "fold", fold_id); rep_df.insert(1, "model", name)
        rep_df.to_csv(f"{OUT_DIR}/perclass_fold{fold_id}_{name}.csv", index=False)
        perclass_rows.append(rep_df)

        # --- (3) 寄与センサ（Permutation Importance 上位20可視化） ---
        try:
            pim = permutation_importance(clf, X_te, y_te, n_repeats=10, random_state=SEED, n_jobs=-1, scoring="f1_macro")
            imp_df = (pd.DataFrame({"feature": feat_names,
                                    "importance": pim.importances_mean,
                                    "std": pim.importances_std})
                      .sort_values("importance", ascending=False))
            imp_df.to_csv(f"{OUT_DIR}/perm_importance_fold{fold_id}_{name}.csv", index=False)

            top = imp_df.head(20)
            plt.figure(figsize=(7.2,5.2))
            plt.barh(top["feature"][::-1], top["importance"][::-1])
            plt.title(f"Permutation Importance (Top 20) - {name} fold {fold_id}")
            plt.tight_layout(); plt.savefig(f"{OUT_DIR}/permimp_top20_{name}_f{fold_id}.png", dpi=150); plt.close()
        except Exception as e:
            print(f"[fold {fold_id} / {name}] Permutation importance skipped:", e)

        # --- (4) PR/ROC と macro AP/AUC（One-vs-Rest） ---
        try:
            proba = clf.predict_proba(X_te)
            y_bin = label_binarize(y_te, classes=np.arange(K))
            # PR
            plt.figure(figsize=(7.2,5.2))
            for i, cls in enumerate(CLASSES):
                PrecisionRecallDisplay.from_predictions(y_bin[:, i], proba[:, i], name=cls)
            plt.title(f"PR Curves (OvR) - {name} fold {fold_id}")
            plt.tight_layout(); plt.savefig(f"{OUT_DIR}/PR_{name}_f{fold_id}.png", dpi=150); plt.close()
            # ROC
            plt.figure(figsize=(7.2,5.2))
            for i, cls in enumerate(CLASSES):
                RocCurveDisplay.from_predictions(y_bin[:, i], proba[:, i], name=cls)
            plt.title(f"ROC Curves (OvR) - {name} fold {fold_id}")
            plt.tight_layout(); plt.savefig(f"{OUT_DIR}/ROC_{name}_f{fold_id}.png", dpi=150); plt.close()
            # macro AP / macro AUC
            ap_macro  = float(np.mean([average_precision_score(y_bin[:,i], proba[:,i]) for i in range(K)]))
            auc_macro = float(np.mean([roc_auc_score(y_bin[:,i], proba[:,i]) for i in range(K)]))
            with open(f"{OUT_DIR}/ovr_macro_{name}_f{fold_id}.txt","w") as f:
                f.write(f"macro_AP={ap_macro:.6f}\nmacro_AUC={auc_macro:.6f}\n")
        except Exception as e:
            print(f"[fold {fold_id} / {name}] PR/ROC skipped:", e)

# ===== (1) 表（%・有効数字4桁）を作成・保存 =====
res_df = pd.DataFrame(all_rows)
res_df.to_csv(f"{OUT_DIR}/cv_raw_metrics.csv", index=False)

mean_df = (res_df.groupby("model")[["accuracy","precision_weighted","recall_weighted",
                                    "f1_weighted","precision_macro","recall_macro","f1_macro"]]
           .mean().reset_index())

show = mean_df.copy()
for c in ["accuracy","precision_weighted","recall_weighted","f1_weighted",
          "precision_macro","recall_macro","f1_macro"]:
    show[c] = show[c].map(pct_4sf)

show.to_csv(f"{OUT_DIR}/cv_summary_percent.csv", index=False)
print("\n=== CV Summary (mean over folds, percent; 4 sig figs) ===")
print(show)
print("\nArtifacts saved in:", OUT_DIR)
```
