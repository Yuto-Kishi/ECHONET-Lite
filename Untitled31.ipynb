```python
# ==============================================================================
# 1. ライブラリのインポート
# ==============================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Google Colabでファイルをアップロードするためのライブラリ
from google.colab import files

# ==============================================================================
# 2. データのアップロードと読み込み
# ==============================================================================
print("ファイル '0919Living_labeled.csv' をアップロードしてください。")
uploaded = files.upload()

# アップロードされたファイル名を取得
file_name = next(iter(uploaded))

print(f"\n'{file_name}' の読み込み中...")
data = pd.read_csv(file_name)
print("読み込み完了。")

# ==============================================================================
# 3. データの前処理と特徴量エンジニアリング（修正版）
# ==============================================================================
print("\nデータの前処理と特徴量エンジニアリングを開始します...")

# タイムスタンプをdatetime型に変換
data['timestamp'] = pd.to_datetime(data['timestamp'])
data = data.sort_values('timestamp') # ★念のため時間でソート

# 時系列データで欠損値がある場合、直前の値で埋める (forward fill)
data.ffill(inplace=True)

# --- 特徴量の選択 ---
#【修正点①】答えに近すぎる 'co2_source' を特徴量から除外
features_to_use = [
    'Living_West__co2', 'M5Stack1__scd40_co2', 'Kitchen__co2',
    'Living_West__temperature', 'M5Stack1__scd40_temp', 'Kitchen__temperature',
    'Living_West__humidity', 'M5Stack1__scd40_hum', 'Kitchen__humidity',
    'M5Stack1__sen55_voc',
    'co2_slope_max'
]
target = 'label_room'

# --- 新しい特徴量の作成：センサー間のCO2濃度の差 ---
data['CO2_Kit_vs_LW'] = data['Kitchen__co2'] - data['Living_West__co2']
data['CO2_Kit_vs_LE'] = data['Kitchen__co2'] - data['M5Stack1__scd40_co2']
data['CO2_LW_vs_LE'] = data['Living_West__co2'] - data['M5Stack1__scd40_co2']
features_to_use.extend(['CO2_Kit_vs_LW', 'CO2_Kit_vs_LE', 'CO2_LW_vs_LE'])

# --- モデル学習用のデータ準備 ---
valid_labels = ['living_west', 'living_east', 'kitchen']
df_model = data[data[target].isin(valid_labels)].copy()
df_model.dropna(subset=features_to_use, inplace=True)
print(f"モデル学習に使用するデータポイント数: {len(df_model)}")

# --- 変数の準備 ---
X = df_model[features_to_use]
le = LabelEncoder()
y = le.fit_transform(df_model[target])
label_mapping = {i: label for i, label in enumerate(le.classes_)}
print("\n部屋ラベルのマッピング:")
print(label_mapping)

# ==============================================================================
# 4. モデルの学習と評価（修正版）
# ==============================================================================
print("\nモデルの学習を開始します...")

#【修正点②】データを時間で分割 (Time-based Split)
# 全データの70%を訓練、残りの30%をテストに使用
split_index = int(len(X) * 0.7)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(f"訓練データ数: {len(X_train)}")
print(f"テストデータ数: {len(X_test)}")

# ランダムフォレストモデルの初期化と学習
model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)
print("学習完了。")

# --- テストデータで性能を評価 ---
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"\n【修正版】モデルの正解率 (Accuracy): {accuracy:.4f}")

# ==============================================================================
# 5. 結果の可視化
# ==============================================================================
print("\n評価結果を可視化します...")

# --- 混同行列 (Confusion Matrix) ---
cm = confusion_matrix(y_test, y_pred, labels=np.arange(len(label_mapping)))
cm_df = pd.DataFrame(cm,
                     index=[label_mapping[i] for i in range(len(label_mapping))],
                     columns=[label_mapping[i] for i in range(len(label_mapping))])

plt.figure(figsize=(10, 7))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix (混同行列) - 修正版', fontsize=16)
plt.ylabel('正解の部屋 (Actual)', fontsize=12)
plt.xlabel('モデルの予測 (Predicted)', fontsize=12)
plt.show()

# --- 特徴量の重要度 (Feature Importance) ---
importances = model.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False).head(15)

plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance (特徴量の重要度 Top 15) - 修正版', fontsize=16)
plt.xlabel('重要度', fontsize=12)
plt.ylabel('特徴量', fontsize=12)
plt.tight_layout()
plt.show()

print("\n--- 処理完了 ---")
```

```python
# ==============================================================================
# 1. ライブラリのインポート
# ==============================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Google Colabでファイルをアップロードするためのライブラリ
from google.colab import files

# ==============================================================================
# 2. データのアップロードと読み込み
# ==============================================================================
print("ファイル '0919Living_labeled.csv' をアップロードしてください。")
uploaded = files.upload()

# アップロードされたファイル名を取得
file_name = next(iter(uploaded))

print(f"\n'{file_name}' の読み込み中...")
data = pd.read_csv(file_name)
print("読み込み完了。")

# ==============================================================================
# 3. データの前処理と特徴量エンジニアリング（再改善版）
# ==============================================================================
print("\nデータの前処理と特徴量エンジニアリングを開始します...")

# タイムスタンプをdatetime型に変換し、インデックスに設定
data['timestamp'] = pd.to_datetime(data['timestamp'])
data = data.sort_values('timestamp').set_index('timestamp')

# --- 特徴量の選択 ---
base_features = [
    'Living_West__co2', 'M5Stack1__scd40_co2', 'Kitchen__co2',
    'Living_West__temperature', 'M5Stack1__scd40_temp', 'Kitchen__temperature',
    'Living_West__humidity', 'M5Stack1__scd40_hum', 'Kitchen__humidity',
    'M5Stack1__sen55_voc',
    'co2_slope_max'
]
target = 'label_room'

# --- 時間的文脈を追加する特徴量エンジニアリング ---
df_features = data[base_features].copy()

# 移動平均 (3分窓)
for col in ['Living_West__co2', 'M5Stack1__scd40_co2', 'Kitchen__co2', 'M5Stack1__sen55_voc']:
    df_features[f'{col}_rolling_mean_3T'] = df_features[col].rolling(window='3T').mean()

# ラグ特徴量 (5分前)
for col in ['Living_West__co2', 'M5Stack1__scd40_co2', 'Kitchen__co2']:
    df_features[f'{col}_lag_5T'] = df_features[col].shift(freq='5T')

# 差分特徴量
df_features['CO2_Kit_vs_LW'] = df_features['Kitchen__co2'] - df_features['Living_West__co2']
df_features['CO2_Kit_vs_LE'] = df_features['Kitchen__co2'] - df_features['M5Stack1__scd40_co2']
df_features['CO2_LW_vs_LE'] = df_features['Living_West__co2'] - df_features['M5Stack1__scd40_co2']

# --- モデル学習用のデータ準備 ---
# 特徴量と目的変数を結合
df_model = df_features.join(data[target])
valid_labels = ['living_west', 'living_east', 'kitchen']
df_model = df_model[df_model[target].isin(valid_labels)].copy()

# 移動平均やラグで発生した最初の欠損値を埋める
df_model.ffill(inplace=True)
# それでも残っている欠損値があれば行ごと削除
df_model.dropna(inplace=True)

print(f"モデル学習に使用するデータポイント数: {len(df_model)}")

# --- 変数の準備 ---
X = df_model.drop(columns=[target])
le = LabelEncoder()
y = le.fit_transform(df_model[target])
label_mapping = {i: label for i, label in enumerate(le.classes_)}
print("\n部屋ラベルのマッピング:")
print(label_mapping)

# ==============================================================================
# 4. モデルの学習と評価（再改善版）
# ==============================================================================
print("\nモデルの学習を開始します...")

# データを時間で分割
split_index = int(len(X) * 0.7)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(f"訓練データ数: {len(X_train)}")
print(f"テストデータ数: {len(X_test)}")

# ランダムフォレストモデルの初期化と学習
# クラスの不均衡を考慮する `class_weight='balanced'` を追加
model = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1, class_weight='balanced')
model.fit(X_train, y_train)
print("学習完了。")

# --- テストデータで性能を評価 ---
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"\n【再改善版】モデルの正解率 (Accuracy): {accuracy:.4f}")

# ==============================================================================
# 5. 結果の可視化
# ==============================================================================
print("\n評価結果を可視化します...")

# --- 混同行列 (Confusion Matrix) ---
cm = confusion_matrix(y_test, y_pred, labels=np.arange(len(label_mapping)))
cm_df = pd.DataFrame(cm,
                     index=[label_mapping[i] for i in range(len(label_mapping))],
                     columns=[label_mapping[i] for i in range(len(label_mapping))])

plt.figure(figsize=(10, 7))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix (混同行列) - 再改善版', fontsize=16)
plt.ylabel('正解の部屋 (Actual)', fontsize=12)
plt.xlabel('モデルの予測 (Predicted)', fontsize=12)
plt.show()

# --- 特徴量の重要度 (Feature Importance) ---
importances = model.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False).head(20) # 上位20

plt.figure(figsize=(12, 10))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df)
plt.title('Feature Importance (特徴量の重要度 Top 20) - 再改善版', fontsize=16)
plt.xlabel('重要度', fontsize=12)
plt.ylabel('特徴量', fontsize=12)
plt.tight_layout()
plt.show()

print("\n--- 処理完了 ---")
```

```python
# ==============================================================================
# 1. ライブラリのインポートとデータ読み込み
# ==============================================================================
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Google Colabでファイルをアップロードするためのライブラリ
from google.colab import files

print("ファイル '0919Living_labeled.csv' をアップロードしてください。")
uploaded = files.upload()
file_name = next(iter(uploaded))
print(f"\n'{file_name}' の読み込み中...")
data = pd.read_csv(file_name)
print("読み込み完了。")

# ==============================================================================
# 2. グラフ作成のためのデータ準備
# ==============================================================================
print("\nグラフ作成の準備をしています...")

# タイムスタンプをdatetime型に変換し、インデックスに設定
data['timestamp'] = pd.to_datetime(data['timestamp'])
data.set_index('timestamp', inplace=True)

# グラフ化する列を選択
co2_columns = {
    'Living_West__co2': '居間 西側 (Living West)',
    'M5Stack1__scd40_co2': '居間 東側 (Living East)',
    'Kitchen__co2': '台所 (Kitchen)'
}

# 描画用にデータをコピーし、列名を分かりやすい日本語に変更
df_plot = data[list(co2_columns.keys())].copy()
df_plot.rename(columns=co2_columns, inplace=True)

# センサーデータは欠損値を含むことがあるため、直前の値で補完
df_plot.ffill(inplace=True)

# ==============================================================================
# 3. グラフの描画と保存
# ==============================================================================
print("グラフを描画しています...")

# グラフのサイズやスタイルを設定
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(15, 7))

# 3つのセンサーデータをプロット
df_plot.plot(ax=ax)

# グラフのタイトルと軸ラベルを設定
ax.set_title('1階 CO₂センサーの時系列推移', fontsize=16)
ax.set_xlabel('時刻', fontsize=12)
ax.set_ylabel('CO₂濃度 (ppm)', fontsize=12)

# 凡例（どの線がどのセンサーかを示すラベル）を設定
ax.legend(title='センサー設置場所', fontsize=10)

# X軸（時刻）のフォーマットを調整して見やすくする
ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
fig.autofmt_xdate() # 日付ラベルを斜めにして重なりを防ぐ

# レイアウトを調整
plt.tight_layout()

# グラフを画像ファイルとして保存
plt.savefig('co2_sensors_timeseries.png')

# グラフを表示
plt.show()

print("\n--- 処理完了 ---")
print("グラフは 'co2_sensors_timeseries.png' という名前で保存されました。")
```
